```python
# @title **Ejemplo 1. construye la grafica de distribucion normal con media 10 y desviacion 2**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
mu = 10  # Media
sigma = 2  # Desviación estándar

# Rango de valores para la gráfica
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)

# Función de densidad de probabilidad (PDF)
y = norm.pdf(x, mu, sigma)

# Crear la gráfica
plt.plot(x, y)
plt.title('Distribución Normal')
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.grid(True)
plt.show()
```

```python
# @title **Ejemplo 2. construye un grafico donde me dibujes las curvas normales con media 10 y las diferentes desviaciones**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
mu = 10  # Media
sigmas = [1, 2, 3]  # Diferentes desviaciones estándar

# Rango de valores para la gráfica
x = np.linspace(mu - 4 * max(sigmas), mu + 4 * max(sigmas), 100)

# Crear la gráfica para cada desviación estándar
for sigma in sigmas:
  y = norm.pdf(x, mu, sigma)
  plt.plot(x, y, label=f'σ = {sigma}')

plt.title('Distribución Normal con diferentes desviaciones estándar')
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.grid(True)
plt.legend() # Agregar leyenda
plt.show()
```

```python
# @title **Ejemplo 3. construye un grafico donde me dibujes diferentes curvas normales con media 8, 10, 12 y 15 y una desviacion de 4**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Medias y desviación estándar
medias = [8, 10, 12, 15]
sigma = 4

# Rango de valores para la gráfica (ajustado para abarcar todas las medias)
x = np.linspace(min(medias) - 4 * sigma, max(medias) + 4 * sigma, 100)

# Crear la gráfica para cada media
for mu in medias:
  y = norm.pdf(x, mu, sigma)
  plt.plot(x, y, label=f'μ = {mu}, σ = {sigma}')

plt.title('Distribución Normal con diferentes medias')
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# @title **Ejemplo 4. Dibuja tres curvas normales con media 21 y desviacion 3 y muestrame en ellas el porcentaje correspondiente a una desviacion estandar, a dos desviacion estandard y a a tres desviacion estandard y escribe el porcentaje en grafico**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
mu = 21  # Media
sigma = 3  # Desviación estándar

# Rango de valores para la gráfica
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)

# Crear la gráfica para cada desviación estándar
for i in range(1, 4):
    y = norm.pdf(x, mu, sigma)
    plt.plot(x, y, label=f'μ = {mu}, σ = {sigma}')

    # Calcular los puntos de desviación estándar
    x1 = mu - sigma * i
    x2 = mu + sigma * i

    # sombrear el área bajo la curva
    plt.fill_between(x[(x > x1) & (x < x2)], 0, norm.pdf(x[(x > x1) & (x < x2)], mu, sigma), alpha=0.3)

    # Calcular el porcentaje dentro de la desviacion
    percentage = round(norm.cdf(x2, mu, sigma) - norm.cdf(x1, mu, sigma),4) * 100

    # Agregar texto al grafico
    plt.text((x1 + x2) / 2, norm.pdf((x1 + x2) / 2, mu, sigma)/2, f'{percentage:.2f}%', ha='center', va='center')

plt.title('tres curvas normales con media 21 y desviacion estandar 3')
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.grid(True)
plt.legend() # Agregar leyenda

plt.show()
```

```python
# @title **Ejercicio 1. resuelve: El volumen de hojas que una oficina de abogados deposita en un juzgados de sus casos atendidos es una variable aleatoria normal para la cual μ = 1200  y σ = 250. Determine la probabilidad de que el volumen depositado sea menor a 1050 hojas y mayor a 980. Haz el grafico, sombrea el area y coloca el porcentaje en el grafico**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
mu = 1200  # Media
sigma = 250  # Desviación estándar

# Rango de valores para la gráfica
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)

# Función de densidad de probabilidad (PDF)
y = norm.pdf(x, mu, sigma)

# Crear la gráfica
plt.plot(x, y)
plt.title('Distribución Normal del Volumen de Hojas')
plt.xlabel('Volumen de Hojas')
plt.ylabel('Probabilidad')
plt.grid(True)

# Límites para el área sombreada
x1 = 980
x2 = 1050

# Sombrear el área bajo la curva entre x1 y x2
plt.fill_between(x[(x > x1) & (x < x2)], 0, norm.pdf(x[(x > x1) & (x < x2)], mu, sigma), alpha=0.5, color='skyblue')

# Calcular la probabilidad
probability = norm.cdf(x2, mu, sigma) - norm.cdf(x1, mu, sigma)

# Agregar texto con el porcentaje
plt.text((x1 + x2) / 2, norm.pdf((x1 + x2) / 2, mu, sigma) / 2, f'{probability * 100:.2f}%', ha='center', va='center')

# Agregar líneas verticales en x1 y x2
plt.axvline(x=x1, color='red', linestyle='--')
plt.axvline(x=x2, color='red', linestyle='--')

plt.show()
```

```python
# @title **Ejercicio 2. resuelve: Sabemos que los paquetes de 1 libra de carne para hamburguesas tienen un contenido medio de 1.00 lb. con una desviación estándar de 0.13, Si tomamos al azar 50 paquetes de 1 libra de carne para hamburguesas calcule la probabilidad de que el peso medio de los paquetes sea mayor a 1.05 libras. Haz el grafico, sombrea el area y coloca el porcentaje en el grafico**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros
mu = 1.00  # Media del peso de los paquetes
sigma = 0.13  # Desviación estándar del peso de los paquetes
n = 50  # Número de paquetes
x_bar = 1.05  # Peso medio a considerar

# Distribución muestral de la media
mu_x_bar = mu
sigma_x_bar = sigma / np.sqrt(n)

# Rango de valores para la gráfica
x = np.linspace(mu_x_bar - 4 * sigma_x_bar, mu_x_bar + 4 * sigma_x_bar, 1000)

# Función de densidad de probabilidad (PDF) de la distribución muestral
y = norm.pdf(x, mu_x_bar, sigma_x_bar)

# Crear la gráfica
plt.plot(x, y)
plt.title('Distribución Muestral del Peso Medio de los Paquetes')
plt.xlabel('Peso Medio de los Paquetes (lb)')
plt.ylabel('Probabilidad')
plt.grid(True)

# Sombrear el área a la derecha de x_bar
x_fill = np.linspace(x_bar, mu_x_bar + 4 * sigma_x_bar, 1000)
plt.fill_between(x_fill, 0, norm.pdf(x_fill, mu_x_bar, sigma_x_bar), alpha=0.3, color='skyblue')

# Calcular la probabilidad
probability = 1 - norm.cdf(x_bar, mu_x_bar, sigma_x_bar)

# Agregar texto con el porcentaje
plt.text(x_bar + 0.01, 0.5, f'{probability * 100:.2f}%', fontsize=12, color='red')

# Agregar una línea vertical en x_bar
plt.axvline(x=x_bar, color='red', linestyle='--')

plt.show()
```

```python
# @title **Ejemplo 1: Construye la grafica de la distribucion normal. con media 10 y desviacion 2**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 10
desviacion_estandar = 12

# Generar datos para la gráfica
x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 100)
y = norm.pdf(x, loc=media, scale=desviacion_estandar)

# Crear la gráfica
plt.plot(x, y)
plt.title('Distribución Normal')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```

```python
# @title **Ejemplo 2: Construye un grafico donde me dibujes las curvas normales con media 10 y diferentes desviaciones 2,3,5 y 8**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Media fija
media = 10

# Diferentes desviaciones estándar
desviaciones_estandar = [2, 3, 5, 8]

# Generar datos y graficar para cada desviación estándar
for desviacion_estandar in desviaciones_estandar:
  x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 100)
  y = norm.pdf(x, loc=media, scale=desviacion_estandar)
  plt.plot(x, y, label=f'Desviación Estándar = {desviacion_estandar}')

# Configurar la gráfica
plt.title('Distribución Normal con Diferentes Desviaciones Estándar')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# @title **Solucion ejercicio 1 : Resuelve este problema: El volumen de hojas que una oficina de abogados deposita en un juzgados de sus casos atendidos es una variable aleatoria normal para la cual μ = 1200  y σ = 250. Determine la probabilidad de que el volumen depositado sea menor a 1050 hojas y mayor a 980 Haz el grafico, sombrea el area y coloca el porcentaje en el grafico**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
mu = 1200
sigma = 250

# Valores de x para el gráfico
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 500)

# Función de densidad de probabilidad
y = norm.pdf(x, loc=mu, scale=sigma)

# Crear el gráfico
plt.plot(x, y)
plt.title('Distribución Normal del Volumen de Hojas')
plt.xlabel('Volumen de Hojas')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# Sombrear el área para x < 1050
x_shade_menor = np.linspace(mu - 3 * sigma, 1050, 200)
y_shade_menor = norm.pdf(x_shade_menor, loc=mu, scale=sigma)
plt.fill_between(x_shade_menor, y_shade_menor, color='purple', alpha=0.5)

# Calcular la probabilidad para x < 1050
probabilidad_mayor = norm.cdf(1050, loc=mu, scale=sigma)

# Sombrear el área para x > 980
x_shade_mayor = np.linspace(980, mu + 3 * sigma, 200)
y_shade_mayor = norm.pdf(x_shade_mayor, loc=mu, scale=sigma)
plt.fill_between(x_shade_mayor, y_shade_mayor, color='lightcoral', alpha=0.5)

# Calcular la probabilidad para x > 980
probabilidad_menor  = norm.cdf(980, loc=mu, scale=sigma)

# Calcular la probabilidad de que el volumen sea menor a 1050 y mayor a 980
probabilidad_total =probabilidad_mayor  - probabilidad_menor

# Agregar texto con las probabilidades en el gráfico
plt.text(900, 0.001, f'P(X < 1050) = {probabilidad_menor*100:.2f}%', fontsize=10)

plt.text(900, 0.0008, f'P(X < 980) = {probabilidad_mayor*100:.2f}%', fontsize=10)

plt.text(900, 0.0006, f'P(980 <X < 1050) = {(probabilidad_mayor-probabilidad_menor)*100:.2f}%', fontsize=10)
# Mostrar el gráfico
plt.show()

print(f"La probabilidad de que el volumen depositado sea menor a 1050 hojas es: {probabilidad_mayor*100:.2f}%")
print(f"La probabilidad de que el volumen depositado sea mayor a 980 hojas es: {probabilidad_menor*100:.2f}%")
print(f"La probabilidad de que el volumen sea menor a 1050 y mayor a 980 es: {probabilidad_total*100:.2f}%")
```

```python
 # @title **Solcuion Ejercicio 2: Resuelve como el problema anterior: Sabemos que los paquetes de 1 libra de carne para hamburguesas tienen un contenido medio de 1.00 lb. con una desviación estándar de 0.13, Si tomamos al azar 50 paquetes de 1 libra de carne para hamburguesas calcule la probabilidad de que el peso medio de los paquetes sea mayor a 1.05 libras , Haz el grafico, sombrea el area y coloca el porcentaje en el grafico**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros del problema
mu = 1.00  # Media del peso de los paquetes
sigma = 0.13  # Desviación estándar del peso de los paquetes
n = 50  # Número de paquetes muestreados
x_bar = 1.05  # Peso medio de los paquetes

# Calcular la desviación estándar de la media muestral
sigma_x_bar = sigma / np.sqrt(n)

# Calcular la probabilidad de que la media muestral sea mayor a 1.05
probabilidad = 1 - norm.cdf(x_bar, loc=mu, scale=sigma_x_bar)

# Generar datos para la gráfica
x = np.linspace(mu - 3 * sigma_x_bar, mu + 3 * sigma_x_bar, 500)
y = norm.pdf(x, loc=mu, scale=sigma_x_bar)

# Crear la gráfica
plt.plot(x, y)
plt.title('Distribución Normal del Peso Medio de los Paquetes')
plt.xlabel('Peso Medio (libras)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# Sombrear el área para x_bar > 1.05
x_shade = np.linspace(x_bar, mu + 3 * sigma_x_bar, 200)
y_shade = norm.pdf(x_shade, loc=mu, scale=sigma_x_bar)
plt.fill_between(x_shade, y_shade, color='skyblue', alpha=0.5)

# Agregar texto con la probabilidad en el gráfico
plt.text(x_bar + 0.01, 0.05, f'{probabilidad*100:.2f}%', fontsize=10)

# Mostrar el gráfico
plt.show()

print(f"La probabilidad de que el peso medio de los paquetes sea mayor a 1.05 libras es: {probabilidad*100:.2f}%")
```

```python
# @title **Ejemplo 3 Resuelve el siguiente ejercicio: Una máquina despachadora de nieve está ajustada para servir un promedio de 30 ml por cono. Si la cantidad de nieve está normalmente distribuida con una varianza de 225 ml. ¿Cuál es la probabilidad de de que un cono contenga entre 25 ml y 35 ml? Haz el grafico, sombrea el area y coloca el porcentaje en el grafico**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 30  # ml
varianza = 225  # ml^2
desviacion_estandar = np.sqrt(varianza)  # ml

# Rango de valores para la gráfica
x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 500)

# Función de densidad de probabilidad
y = norm.pdf(x, loc=media, scale=desviacion_estandar)

# Crear la gráfica
plt.plot(x, y)
plt.title('Distribución Normal de la Cantidad de Nieve')
plt.xlabel('Cantidad de Nieve (ml)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# Sombrear el área entre 25 ml y 35 ml
x_shade = np.linspace(25, 35, 200)
y_shade = norm.pdf(x_shade, loc=media, scale=desviacion_estandar)
plt.fill_between(x_shade, y_shade, color='purple', alpha=0.5)

# Calcular la probabilidad
probabilidad = norm.cdf(35, loc=media, scale=desviacion_estandar) - norm.cdf(25, loc=media, scale=desviacion_estandar)

# Agregar texto con la probabilidad en el gráfico
plt.text(26, 0.01, f'{probabilidad*100:.2f}%', fontsize=10)


# Mostrar el gráfico
plt.show()

print(f"La probabilidad de que un cono contenga entre 25 ml y 35 ml de nieve es: {probabilidad*100:.2f}%")
```

```python
# @title **Ejemplo 9: Resuelve: Las edades de los Estudiantes del prof. Hurtado siguen una distribución Normal con Media de 18.7 y desviación de 1.3. Si tomamos una muestra de 25 estudiantes, calcula la Probabilidad de que un estudiante seleccionado al azar tenga una edad promedio mayor a 21 años. Haz el grafico, sombrea el area y coloca el porcentaje en el grafico**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media_poblacional = 18.7
desviacion_estandar_poblacional = 1.3
n = 25  # Tamaño de la muestra
media_muestral = 21

# Calcular la desviación estándar de la media muestral
desviacion_estandar_muestral = desviacion_estandar_poblacional / np.sqrt(n)

# Calcular la probabilidad de que la media muestral sea mayor a 21
probabilidad = 1 - norm.cdf(media_muestral, loc=media_poblacional, scale=desviacion_estandar_muestral)

# Generar datos para la gráfica
x = np.linspace(media_poblacional - 3 * desviacion_estandar_muestral, media_poblacional + 3 * desviacion_estandar_muestral, 500)
y = norm.pdf(x, loc=media_poblacional, scale=desviacion_estandar_muestral)

# Crear la gráfica
plt.plot(x, y)
plt.title('Distribución Normal de la Media Muestral de Edades')
plt.xlabel('Media Muestral de Edades')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# Sombrear el área para x > 21
x_shade = np.linspace(media_muestral, media_poblacional + 3 * desviacion_estandar_muestral, 200)
y_shade = norm.pdf(x_shade, loc=media_poblacional, scale=desviacion_estandar_muestral)
plt.fill_between(x_shade, y_shade, color='purple', alpha=0.5)

# Agregar texto con la probabilidad en el gráfico
plt.text(media_muestral - 0.5, 0.05, f'{probabilidad*100:.2f}%', fontsize=12)

# Mostrar el gráfico
plt.show()

print(f"La probabilidad de que la media muestral de edades sea mayor a 21 años es: {probabilidad*100:.2f}%")
```

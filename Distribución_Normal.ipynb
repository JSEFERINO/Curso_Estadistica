```python
# prompt: Generame una distribucion normal con media 10 y desviacion 2 y graficala

import numpy as np
import matplotlib.pyplot as plt

# Generar una distribución normal con media 10 y desviación estándar 2
media = 10
desviacion_estandar = 2
datos = np.random.normal(loc=media, scale=desviacion_estandar, size=1000)

# Graficar la distribución
plt.hist(datos, bins=30, density=True)
plt.title('Distribución Normal')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# prompt: Dibuja la curva normal encima de ella

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar una distribución normal con media 10 y desviación estándar 2
media = 10
desviacion_estandar = 2
datos = np.random.normal(loc=media, scale=desviacion_estandar, size=1000)

# Graficar la distribución
plt.hist(datos, bins=30, density=True, alpha=0.6, color='g')
plt.title('Distribución Normal')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

# Generar la curva normal teórica
x_values = np.linspace(min(datos), max(datos), 100)
y_values = norm.pdf(x_values, loc=media, scale=desviacion_estandar)

# Graficar la curva normal encima del histograma
plt.plot(x_values, y_values, 'r', linewidth=2)

plt.show()
```

```python
# prompt: Dibuja cinco curvas normales con media 10 y desviaciones 1, 3, 5, 7, 9, 4

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de las curvas
medias = [10] * 5
desviaciones_estandar = [1, 3, 5, 7, 9]

# Generar la gráfica
plt.figure(figsize=(10, 6))

for media, desviacion_estandar in zip(medias, desviaciones_estandar):
  x_values = np.linspace(media - 4 * desviacion_estandar, media + 4 * desviacion_estandar, 100)
  y_values = norm.pdf(x_values, loc=media, scale=desviacion_estandar)
  plt.plot(x_values, y_values, label=f'Desviación Estándar: {desviacion_estandar}')

plt.title('Curvas Normales con Media 10 y Diferentes Desviaciones Estándar')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: Resuelve el problema, haz el grafico, sombrea la region. El volumen de hojas que una oficina de abogados deposita en un juzgados de sus casos atendidos es una variable aleatoria normal para la cual μ = 1200  y σ = 250. Determine la probabilidad de que el volumen depositado sea menor a 1050 hojas y mayor a 980

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 1200
desviacion_estandar = 250

# Valores para los que se calculará la probabilidad
x1 = 1050
x2 = 980

# Calcular la probabilidad de que el volumen sea menor a 1050
probabilidad_menor_1050 = norm.cdf(x1, loc=media, scale=desviacion_estandar)

# Calcular la probabilidad de que el volumen sea menor a 980
probabilidad_menor_980 = norm.cdf(x2, loc=media, scale=desviacion_estandar)

# Calcular la probabilidad de que el volumen sea mayor a 980 y menor a 1050
probabilidad_entre_980_1050 = probabilidad_menor_1050 - probabilidad_menor_980


# Generar la gráfica de la distribución normal
x_values = np.linspace(media - 4 * desviacion_estandar, media + 4 * desviacion_estandar, 100)
y_values = norm.pdf(x_values, loc=media, scale=desviacion_estandar)

plt.plot(x_values, y_values)
plt.title('Distribución Normal del Volumen de Hojas')
plt.xlabel('Volumen de Hojas')
plt.ylabel('Densidad de Probabilidad')

# Sombrear la región entre 980 y 1050
plt.fill_between(x_values, y_values, where=(x_values >= x2) & (x_values <= x1), color='blue', alpha=0.5)

plt.show()

print(f"La probabilidad de que el volumen sea menor a 1050 es: {probabilidad_menor_1050:.5f}")
print(f"La probabilidad de que el volumen sea menor a 980 es: {probabilidad_menor_980:.5f}")
print(f"La probabilidad de que el volumen sea mayor a 980 y menor a 1050 es: {probabilidad_entre_980_1050:.5f}")
```

```python
# prompt: Adapta el siguiente problema: Una máquina despachadora de nieve está ajustada para servir un promedio de 30 ml por cono. Si la cantidad de nieve está normalmente distribuida con una varianza de 225 ml. ¿Cuál es la probabilidad de de que un cono contenga entre 25 ml y 35 ml? Dibuja region y escribe la probabilidad en el grafico

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 30
desviacion_estandar = np.sqrt(225)  # Varianza de 225, desviación estándar es la raíz cuadrada

# Valores para los que se calculará la probabilidad
x1 = 35
x2 = 25

# Calcular la probabilidad de que la cantidad de nieve sea menor a 35 ml
probabilidad_menor_35 = norm.cdf(x1, loc=media, scale=desviacion_estandar)

# Calcular la probabilidad de que la cantidad de nieve sea menor a 25 ml
probabilidad_menor_25 = norm.cdf(x2, loc=media, scale=desviacion_estandar)

# Calcular la probabilidad de que la cantidad de nieve sea mayor a 25 ml y menor a 35 ml
probabilidad_entre_25_35 = probabilidad_menor_35 - probabilidad_menor_25


# Generar la gráfica de la distribución normal
x_values = np.linspace(media - 4 * desviacion_estandar, media + 4 * desviacion_estandar, 100)
y_values = norm.pdf(x_values, loc=media, scale=desviacion_estandar)

plt.plot(x_values, y_values)
plt.title('Distribución Normal de la Cantidad de Nieve en un Cono')
plt.xlabel('Cantidad de Nieve (ml)')
plt.ylabel('Densidad de Probabilidad')

# Sombrear la región entre 25 ml y 35 ml
plt.fill_between(x_values, y_values, where=(x_values >= x2) & (x_values <= x1), color='blue', alpha=0.5)

plt.show()

print(f"La probabilidad de que un cono contenga entre 25 ml y 35 ml es: {probabilidad_entre_25_35:.5f}")
```

```python
# prompt: La Estatura de mis estudiantes siguen una distribucion normal con media 168 y desviacion estandard de 9. Si elegimos un estudiante al azar, cual es la probabilidad de que su estatura se encunetre entre 150 y 170. Adaptalo al ejemplo anterior. con llos graficos y las areas

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 168
desviacion_estandar = 9

# Valores para los que se calculará la probabilidad
x1 = 170
x2 = 150

# Calcular la probabilidad de que la estatura sea menor a 170 cm
probabilidad_menor_170 = norm.cdf(x1, loc=media, scale=desviacion_estandar)

# Calcular la probabilidad de que la estatura sea menor a 150 cm
probabilidad_menor_150 = norm.cdf(x2, loc=media, scale=desviacion_estandar)

# Calcular la probabilidad de que la estatura sea mayor a 150 cm y menor a 170 cm
probabilidad_entre_150_170 = probabilidad_menor_170 - probabilidad_menor_150


# Generar la gráfica de la distribución normal
x_values = np.linspace(media - 4 * desviacion_estandar, media + 4 * desviacion_estandar, 100)
y_values = norm.pdf(x_values, loc=media, scale=desviacion_estandar)

plt.plot(x_values, y_values)
plt.title('Distribución Normal de la Estatura de los Estudiantes')
plt.xlabel('Estatura (cm)')
plt.ylabel('Densidad de Probabilidad')

# Sombrear la región entre 150 cm y 170 cm
plt.fill_between(x_values, y_values, where=(x_values >= x2) & (x_values <= x1), color='blue', alpha=0.5)
plt.text(175, 0.02, f' P(150<X<170)= {probabilidad_entre_150_170:.5f}', ha='center', fontsize=12) # Agregar texto al gráfico

plt.show()
print(f"La probabilidad de que la estatura sea menor a 170 cm es: {probabilidad_menor_170:.5f}")
print(f"La probabilidad de que la estatura sea menor a 150 cm es: {probabilidad_menor_150:.5f}")
print(f"La probabilidad de que la estatura esté entre 150 cm y 170 cm es: {probabilidad_entre_150_170:.5f}")
```

```python
# prompt: La Estatura de mis estudiantes siguen una distribucion normal con media 168 y desviacion estandard de 9. Si elegimos una muestra de 25 estudiante al azar, cual es la probabilidad de que su estatura promedio se encunetre entre 150 y 170. Adaptalo al ejemplo anterior. con llos graficos y las areas

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal de la estatura de los estudiantes
media_poblacional = 168
desviacion_estandar_poblacional = 9
n = 25  # Tamaño de la muestra

# Calcular la desviación estándar de la media muestral
desviacion_estandar_media_muestral = desviacion_estandar_poblacional / np.sqrt(n)

# Valores para los que se calculará la probabilidad
x1 = 170
x2 = 150

# Calcular la probabilidad de que la media muestral sea menor a 170 cm
probabilidad_menor_170 = norm.cdf(x1, loc=media_poblacional, scale=desviacion_estandar_media_muestral)

# Calcular la probabilidad de que la media muestral sea menor a 150 cm
probabilidad_menor_150 = norm.cdf(x2, loc=media_poblacional, scale=desviacion_estandar_media_muestral)

# Calcular la probabilidad de que la media muestral esté entre 150 cm y 170 cm
probabilidad_entre_150_170 = probabilidad_menor_170 - probabilidad_menor_150


# Generar la gráfica de la distribución normal de la media muestral
x_values = np.linspace(media_poblacional - 4 * desviacion_estandar_media_muestral,
                       media_poblacional + 4 * desviacion_estandar_media_muestral, 100)
y_values = norm.pdf(x_values, loc=media_poblacional, scale=desviacion_estandar_media_muestral)

plt.plot(x_values, y_values)
plt.title('Distribución Normal de la Media Muestral de la Estatura')
plt.xlabel('Media Muestral de la Estatura (cm)')
plt.ylabel('Densidad de Probabilidad')

# Sombrear la región entre 150 cm y 170 cm
plt.fill_between(x_values, y_values, where=(x_values >= x2) & (x_values <= x1), color='blue', alpha=0.5)
plt.text(175, 0.05, f'P(150 < X̄ < 170) = {probabilidad_entre_150_170:.5f}', ha='center', fontsize=12)


plt.show()
print(f"La probabilidad de que la media muestral sea menor a 170 cm es: {probabilidad_menor_170:.6f}")
print(f"La probabilidad de que la media muestral sea menor a 150 cm es: {probabilidad_menor_150:.6f}")

print(f"La probabilidad de que la media muestral de la estatura esté entre 150 cm y 170 cm es: {probabilidad_entre_150_170:.6f}")
```

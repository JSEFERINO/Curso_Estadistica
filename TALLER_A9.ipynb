# **Hallar las medidas de tendencia central para datos individuales: Media, Mediana y Moda. ***

```python
# prompt: Medidas de tendencia central de los datos individuales, media, mediana, moda

peso = [45, 56, 67, 69, 50, 61, 50, 59, 70, 73, 65, 65, 57, 70, 80, 70,
          62, 57, 60, 58, 65, 72, 77, 71, 48, 90, 67, 55, 67, 72, 67,
          102, 55, 60, 60, 75, 76, 55, 48, 88, 50, 59, 60, 60, 64,
          53, 49, 70, 80, 42, 59, 64, 65, 80, 62, 55, 60, 53, 62,
          69, 60, 76, 78, 70, 78, 52, 55, 48, 82, 65, 65, 50, 50,
          54, 55, 60]

# Calcular la media
media = sum(peso) / len(peso)

# Calcular la mediana
peso_ordenado = sorted(peso)
n = len(peso_ordenado)
if n % 2 == 0:
  mediana = (peso_ordenado[n//2 - 1] + peso_ordenado[n//2]) / 2
else:
  mediana = peso_ordenado[n//2]

# Calcular la moda
frecuencias = {}
for valor in peso:
  if valor in frecuencias:
    frecuencias[valor] += 1
  else:
    frecuencias[valor] = 1

moda = []
max_frecuencia = 0
for valor, frecuencia in frecuencias.items():
  if frecuencia > max_frecuencia:
    moda = [valor]
    max_frecuencia = frecuencia
  elif frecuencia == max_frecuencia and valor not in moda:
    moda.append(valor)

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda)
```

# **2. Hallar medidas de posicion para los datos individuales: Q1, Q3, D3, D8. ***

```python
# prompt: Medidas de posicion, cuartil 1, cuartil 3, decil 3 y decil 8

peso = [45, 56, 67, 69, 50, 61, 50, 59, 70, 73, 65, 65, 57, 70, 80, 70,
          62, 57, 60, 58, 65, 72, 77, 71, 48, 90, 67, 55, 67, 72, 67,
          102, 55, 60, 60, 75, 76, 55, 48, 88, 50, 59, 60, 60, 64,
          53, 49, 70, 80, 42, 59, 64, 65, 80, 62, 55, 60, 53, 62,
          69, 60, 76, 78, 70, 78, 52, 55, 48, 82, 65, 65, 50, 50,
          54, 55, 60]

peso_ordenado = sorted(peso)
n = len(peso_ordenado)

# Calcular el cuartil 1 (Q1)
posicion_q1 = (n + 1) / 4
if posicion_q1 == int(posicion_q1):
  q1 = peso_ordenado[int(posicion_q1) - 1]
else:
  parte_entera = int(posicion_q1)
  parte_decimal = posicion_q1 - parte_entera
  q1 = peso_ordenado[parte_entera - 1] + parte_decimal * (peso_ordenado[parte_entera] - peso_ordenado[parte_entera - 1])

# Calcular el cuartil 3 (Q3)
posicion_q3 = 3 * (n + 1) / 4
if posicion_q3 == int(posicion_q3):
  q3 = peso_ordenado[int(posicion_q3) - 1]
else:
  parte_entera = int(posicion_q3)
  parte_decimal = posicion_q3 - parte_entera
  q3 = peso_ordenado[parte_entera - 1] + parte_decimal * (peso_ordenado[parte_entera] - peso_ordenado[parte_entera - 1])

# Calcular el decil 3 (D3)
posicion_d3 = 3 * (n + 1) / 10
if posicion_d3 == int(posicion_d3):
  d3 = peso_ordenado[int(posicion_d3) - 1]
else:
  parte_entera = int(posicion_d3)
  parte_decimal = posicion_d3 - parte_entera
  d3 = peso_ordenado[parte_entera - 1] + parte_decimal * (peso_ordenado[parte_entera] - peso_ordenado[parte_entera - 1])

# Calcular el decil 8 (D8)
posicion_d8 = 8 * (n + 1) / 10
if posicion_d8 == int(posicion_d8):
  d8 = peso_ordenado[int(posicion_d8) - 1]
else:
  parte_entera = int(posicion_d8)
  parte_decimal = posicion_d8 - parte_entera
  d8 = peso_ordenado[parte_entera - 1] + parte_decimal * (peso_ordenado[parte_entera] - peso_ordenado[parte_entera - 1])


print("Cuartil 1 (Q1):", q1)
print("Cuartil 3 (Q3):", q3)
print("Decil 3 (D3):", d3)
print("Decil 8 (D8):", d8)
```

# **4.Tabla agrupada de frecuencias.**

```python
# prompt: Tabla de frecuencia agrupada

import pandas as pd

peso = [45, 56, 67, 69, 50, 61, 50, 59, 70, 73, 65, 65, 57, 70, 80, 70,
          62, 57, 60, 58, 65, 72, 77, 71, 48, 90, 67, 55, 67, 72, 67,
          102, 55, 60, 60, 75, 76, 55, 48, 88, 50, 59, 60, 60, 64,
          53, 49, 70, 80, 42, 59, 64, 65, 80, 62, 55, 60, 53, 62,
          69, 60, 76, 78, 70, 78, 52, 55, 48, 82, 65, 65, 50, 50,
          54, 55, 60]

# Definir el número de intervalos (se puede ajustar según la cantidad de datos)
num_intervalos = 7

# Calcular el rango de los datos
rango = max(peso) - min(peso)

# Calcular el ancho del intervalo
ancho_intervalo = rango / num_intervalos

# Crear los intervalos
intervalos = []
inicio_intervalo = min(peso)
for i in range(num_intervalos):
  fin_intervalo = inicio_intervalo + ancho_intervalo
  intervalos.append((inicio_intervalo, fin_intervalo))
  inicio_intervalo = fin_intervalo

# Inicializar las listas para almacenar la información de la tabla
marca_clase = []
frecuencia_absoluta = []
frecuencia_absoluta_acumulada = []
frecuencia_relativa = []
frecuencia_relativa_acumulada = []

frecuencia_acumulada = 0
for intervalo in intervalos:
  # Calcular la marca de clase
  marca = (intervalo[0] + intervalo[1]) / 2
  marca_clase.append(marca)

  # Contar la frecuencia absoluta del intervalo
  frecuencia = sum(1 for peso_i in peso if intervalo[0] <= peso_i < intervalo[1])
  frecuencia_absoluta.append(frecuencia)

  # Calcular la frecuencia absoluta acumulada
  frecuencia_acumulada += frecuencia
  frecuencia_absoluta_acumulada.append(frecuencia_acumulada)

  # Calcular la frecuencia relativa
  frecuencia_relativa.append(frecuencia / len(peso))

  # Calcular la frecuencia relativa acumulada
  frecuencia_relativa_acumulada.append(frecuencia_acumulada / len(peso))

# Crear un DataFrame de Pandas con la tabla de frecuencias
tabla_frecuencias = pd.DataFrame({
    'Intervalo': intervalos,
    'Marca de Clase': marca_clase,
    'Frecuencia Absoluta': frecuencia_absoluta,
    'Frecuencia Absoluta Acumulada': frecuencia_absoluta_acumulada,
    'Frecuencia Relativa': frecuencia_relativa,
    'Frecuencia Relativa Acumulada': frecuencia_relativa_acumulada
})

print(tabla_frecuencias)
```

```python
# prompt: Diagrama de tallo y hoja.

peso = [45, 56, 67, 69, 50, 61, 50, 59, 70, 73, 65, 65, 57, 70, 80, 70,
          62, 57, 60, 58, 65, 72, 77, 71, 48, 90, 67, 55, 67, 72, 67,
          102, 55, 60, 60, 75, 76, 55, 48, 88, 50, 59, 60, 60, 64,
          53, 49, 70, 80, 42, 59, 64, 65, 80, 62, 55, 60, 53, 62,
          69, 60, 76, 78, 70, 78, 52, 55, 48, 82, 65, 65, 50, 50,
          54, 55, 60]

# Diagrama de tallo y hoja
def stem_leaf_plot(data):
    stems = []
    leaves = []
    for value in data:
        stem = value // 10
        leaf = value % 10
        if stem not in stems:
            stems.append(stem)
            leaves.append([leaf])
        else:
            leaves[stems.index(stem)].append(leaf)

    for i in range(len(stems)):
        print(f"{stems[i]} | {' '.join(map(str, sorted(leaves[i])))}")


stem_leaf_plot(peso)
```

```python
# prompt: Medida de posicion para datos agrupados: cuartil 1, cuartil 3, decil 3, decil 8.
import pandas as pd

peso = [45, 56, 67, 69, 50, 61, 50, 59, 70, 73, 65, 65, 57, 70, 80, 70,
          62, 57, 60, 58, 65, 72, 77, 71, 48, 90, 67, 55, 67, 72, 67,
          102, 55, 60, 60, 75, 76, 55, 48, 88, 50, 59, 60, 60, 64,
          53, 49, 70, 80, 42, 59, 64, 65, 80, 62, 55, 60, 53, 62,
          69, 60, 76, 78, 70, 78, 52, 55, 48, 82, 65, 65, 50, 50,
          54, 55, 60]

# Definir el número de intervalos (se puede ajustar según la cantidad de datos)
num_intervalos = 7

# Calcular el rango de los datos
rango = max(peso) - min(peso)

# Calcular el ancho del intervalo
ancho_intervalo = rango / num_intervalos

# Crear los intervalos
intervalos = []
inicio_intervalo = min(peso)
for i in range(num_intervalos):
  fin_intervalo = inicio_intervalo + ancho_intervalo
  intervalos.append((inicio_intervalo, fin_intervalo))
  inicio_intervalo = fin_intervalo

# Inicializar las listas para almacenar la información de la tabla
marca_clase = []
frecuencia_absoluta = []
frecuencia_absoluta_acumulada = []
frecuencia_relativa = []
frecuencia_relativa_acumulada = []

frecuencia_acumulada = 0
for intervalo in intervalos:
  # Calcular la marca de clase
  marca = (intervalo[0] + intervalo[1]) / 2
  marca_clase.append(marca)

  # Contar la frecuencia absoluta del intervalo
  frecuencia = sum(1 for peso_i in peso if intervalo[0] <= peso_i < intervalo[1])
  frecuencia_absoluta.append(frecuencia)

  # Calcular la frecuencia absoluta acumulada
  frecuencia_acumulada += frecuencia
  frecuencia_absoluta_acumulada.append(frecuencia_acumulada)

  # Calcular la frecuencia relativa
  frecuencia_relativa.append(frecuencia / len(peso))

  # Calcular la frecuencia relativa acumulada
  frecuencia_relativa_acumulada.append(frecuencia_acumulada / len(peso))

# Crear un DataFrame de Pandas con la tabla de frecuencias
tabla_frecuencias = pd.DataFrame({
    'Intervalo': intervalos,
    'Marca de Clase': marca_clase,
    'Frecuencia Absoluta': frecuencia_absoluta,
    'Frecuencia Absoluta Acumulada': frecuencia_absoluta_acumulada,
    'Frecuencia Relativa': frecuencia_relativa,
    'Frecuencia Relativa Acumulada': frecuencia_relativa_acumulada
})

# Calcular el cuartil 1 (Q1) agrupado
n = len(peso)
posicion_q1 = (n + 1) / 4
for index, row in tabla_frecuencias.iterrows():
  if row['Frecuencia Absoluta Acumulada'] >= posicion_q1:
    q1_agrupado = row['Marca de Clase']
    break

print("Cuartil 1 (Q1) agrupado:", q1_agrupado)

# Calcular el cuartil 3 (Q3) agrupado
posicion_q3 = 3 * (n + 1) / 4
for index, row in tabla_frecuencias.iterrows():
  if row['Frecuencia Absoluta Acumulada'] >= posicion_q3:
    q3_agrupado = row['Marca de Clase']
    break

print("Cuartil 3 (Q3) agrupado:", q3_agrupado)


# Calcular el decil 3 (D3) agrupado
posicion_d3 = 3 * (n + 1) / 10
for index, row in tabla_frecuencias.iterrows():
  if row['Frecuencia Absoluta Acumulada'] >= posicion_d3:
    d3_agrupado = row['Marca de Clase']
    break

print("Decil 3 (D3) agrupado:", d3_agrupado)


# Calcular el decil 8 (D8) agrupado
posicion_d8 = 8 * (n + 1) / 10
for index, row in tabla_frecuencias.iterrows():
  if row['Frecuencia Absoluta Acumulada'] >= posicion_d8:
    d8_agrupado = row['Marca de Clase']
    break

print("Decil 8 (D8) agrupado:", d8_agrupado)
```

# *** 10. Intervalo de confianza de 95% para la media real de la nota definitiva.**

```python
# prompt: Intervalo de confianza de 95% para la media real de la nota definitiva.

import numpy as np
from scipy import stats

peso = [45, 56, 67, 69, 50, 61, 50, 59, 70, 73, 65, 65, 57, 70, 80, 70,
          62, 57, 60, 58, 65, 72, 77, 71, 48, 90, 67, 55, 67, 72, 67,
          102, 55, 60, 60, 75, 76, 55, 48, 88, 50, 59, 60, 60, 64,
          53, 49, 70, 80, 42, 59, 64, 65, 80, 62, 55, 60, 53, 62,
          69, 60, 76, 78, 70, 78, 52, 55, 48, 82, 65, 65, 50, 50,
          54, 55, 60]

# Calcular la media muestral
media_muestral = np.mean(peso)

# Calcular la desviación estándar muestral
desviacion_estandar_muestral = np.std(peso, ddof=1)  # ddof=1 para la desviación estándar muestral

# Calcular el error estándar
n = len(peso)
error_estandar = desviacion_estandar_muestral / np.sqrt(n)

# Calcular el intervalo de confianza del 95%
intervalo_confianza = stats.t.interval(0.95, df=n-1, loc=media_muestral, scale=error_estandar)

print("Intervalo de confianza del 95% para la media real de la nota definitiva:", intervalo_confianza)
```

# **Hay un 5% de probabilidad de que el intervalo calculado no contenga la media real de la población. **

# **Ejercicio 1.**
+ Un profesor de la UTB nunca termina su clase antes del
término de la hora, más nunca se pasa de 2 minutos de ésta. Sea X : el tiempo que transcurre entre el término de la hora y el término efectivo de la clase. Suponga que la fdp de X viene dada por:

$$ f(x) = \begin{cases}
kx^2 & \text{si } 0 \leq x \leq 10 \\
0 & \text{en otro caso}
\end{cases}$$

 + Encuentre el valor de k.
 + ¿Cuál es la probabilidad de que la clase termine a menos de un minuto después del término de la hora?
 + ¿Cuál es la probabilidad de que la clase continúe entre 60 y 90 segundos después del término de la hora?
 + ¿Cuál es la probabilidad de que la clase continúe por lo menos 90 segundos después del término de la hora?


# **Solución del Ejercicio 1.**

+ Para resolver este ejercicio, primero determinemos el valor de $k$ utilizando la función de densidad de probabilidad (fdp) proporcionada:

$ f(x) = \begin{cases}
kx^2 & \text{si } 0 \leq x \leq 10 \\
0 & \text{en otro caso}
\end{cases} $

+ **Paso 1: Encontrar \( k \)**

  + La integral de la fdp sobre todos los posibles valores de \( x \) debe ser igual a 1 (ya que es una función de densidad de probabilidad).

$$\int_{0}^{10} kx^2 \, dx = 1 $$

+ Calculamos la integral:

$$\int_{0}^{10} kx^2 \, dx = k \left[ \frac{x^3}{3} \right]_{0}^{10} = k \left( \frac{10^3}{3} - \frac{0^3}{3} \right) = k \cdot \frac{1000}{3} = \frac{1000k}{3}$$

+ Igualando a 1:

$$\frac{1000k}{3} = 1$$

$$1000k = 3$$

$$k = \frac{3}{1000} = 0.003$$

+ Entonces, $k = 0.003$.

+ **Paso 2: Probabilidad de que la clase termine a menos de un minuto después del término de la hora (\( X < 1 \))**

$$P(X < 1) = \int_{0}^{1} 0.003x^2 \, dx $$

+ Calculamos la integral:

$$P(X < 1) = 0.003 \cdot \left[ \frac{x^3}{3} \right]_{0}^{1} = 0.003 \cdot \frac{1^3}{3} = 0.003 \cdot \frac{1}{3} = 0.001$$

+ **Paso 3: Probabilidad de que la clase continúe entre 60 y 90 segundos después del término de la hora $ 1 \leq X \leq 1.5$**

$$P(1 \leq X \leq 1.5) = \int_{1}^{1.5} 0.003x^2 \, dx$$

+ Calculamos la integral:

$$P(1 \leq X \leq 1.5) = 0.003 \cdot \left[ \frac{x^3}{3} \right]_{1}^{1.5} = 0.003 \cdot \left( \frac{(1.5)^3}{3} - \frac{1^3}{3} \right)$$

$$P(1 \leq X \leq 1.5) = 0.003 \cdot \left( \frac{3.375 - 1}{3} \right) = 0.003 \cdot \frac{2.375}{3} \approx 0.002375$$

+ **Paso 4: Probabilidad de que la clase continúe por lo menos 90 segundos después del término de la hora $X \geq 1.5$**

$$P(X \geq 1.5) = \int_{1.5}^{10} 0.003x^2 \, dx$$

+ Calculamos la integral:

$$P(X \geq 1.5) = 0.003 \cdot \left[ \frac{x^3}{3} \right]_{1.5}^{10} = 0.003 \cdot \left( \frac{10^3}{3} - \frac{(1.5)^3}{3} \right)$$

$$P(X \geq 1.5) = 0.003 \cdot \left( \frac{1000}{3} - \frac{3.375}{3} \right) = 0.003 \cdot \left( \frac{996.625}{3} \right) \approx 0.998875$$

+ Estas son las respuestas a las tres preguntas planteadas sobre la distribución del tiempo $X$ entre el término de la hora y el término efectivo de la clase.

# **Solución del Ejercicio 1. - Usando Python**

```python
#!pip install scipy &>/dev/null
import scipy.integrate as integrate

# Definir la función de densidad de probabilidad f(x)
def f(x):
    if 0 <= x <= 10:
        return 0.003 * x**2
    else:
        return 0

# Calcular la constante de normalización k
integral, _ = integrate.quad(f, 0, 10)  # Calcula la integral de f(x) de 0 a 10
k = 1 / integral  # Normalización para que la integral de f(x) de 0 a 10 sea 1
print(f"Valor de k: {k}")

# Función de densidad de probabilidad normalizada
def f_normalized(x):
    return k * f(x)

# Probabilidad de que la clase termine a menos de un minuto después del término de la hora (X < 1)
prob_less_than_1, _ = integrate.quad(f_normalized, 0, 1)
print(f"Probabilidad pedida P(X < 1) = {prob_less_than_1:.4f}")

# Probabilidad de que la clase continúe entre 60 y 90 segundos después del término de la hora (1 <= X <= 1.5)
prob_between_1_and_1_5, _ = integrate.quad(f_normalized, 1, 1.5)
print(f"Probabilidad pedida P(1 <= X <= 1.5) = {prob_between_1_and_1_5:.4f}")

# Probabilidad de que la clase continúe por lo menos 90 segundos después del término de la hora (X >= 1.5)
prob_at_least_1_5, _ = integrate.quad(f_normalized, 1.5, 10)
print(f"Probabilidad pedida P(X >= 1.5) =  {prob_at_least_1_5:.4f}")
```

# **Ejercicio 2.**
+   La distribución de la cantidad de grava (en toneladas)
vendida a una empresa en particular proveedora de materiales
para la construcción, en una semana dada, es una v.a. X continua
con fdp:
$$f(x) = \begin{cases}
\frac{3}{2}(1 - x^2) & \text{si } 0 \leq x \leq 1 \\
0 & \text{en otro caso}
\end{cases}$$

+ **a**. ¿Cuántas toneladas esperarías que se vendan durante esa semana?
+ **b**. Halle la varianza y la desviación estándar de X


# **Solución del Ejercicio 2.**
+ Para resolver este ejercicio, primero debemos trabajar con la función de densidad de probabilidad (fdp) proporcionada:

$$f(x) = \begin{cases}
\frac{3}{2}(1 - x^2) & \text{si } 0 \leq x \leq 1 \\
0 & \text{en otro caso}
\end{cases}$$

+ **Paso 1. Calcular el valor esperado (esperanza)**

+ La esperanza $E[X]$ de una variable aleatoria continua con fdp $f(x)$ se calcula como:

$$E[X] = \int_{-\infty}^{\infty} x f(x) \, dx$$

+ Dado que $f(x)$ es no nula solo en \( 0 \leq x \leq 1 \), la esperanza se calcula como:

$$E[X] = \int_{0}^{1} x \cdot \frac{3}{2}(1 - x^2) \, dx$$

+ Calculamos esta integral:

$$E[X] = \frac{3}{2} \int_{0}^{1} x (1 - x^2) \, dx$$

$$E[X] = \frac{3}{2} \left( \int_{0}^{1} x \, dx - \int_{0}^{1} x^3 \, dx \right)$$

$$E[X] = \frac{3}{2} \left( \left[ \frac{x^2}{2} \right]_{0}^{1} - \left[ \frac{x^4}{4} \right]_{0}^{1} \right)$$

$$E[X] = \frac{3}{2} \left( \frac{1}{2} - \frac{1}{4} \right)$$

$$E[X] = \frac{3}{2} \cdot \frac{1}{4}$$

$$E[X] = \frac{3}{8}$$

+ Por lo tanto, esperamos que se vendan \( \frac{3}{8} \) toneladas de grava durante esa semana.

+ **2. Calcular la varianza y la desviación estándar**

+ La varianza $\text{Var}(X)$ de una variable aleatoria continua se calcula como:

$$\text{Var}(X) = E[X^2] - (E[X])^2$$

+ **a. Calcular $E[X^2]$**

$$E[X^2] = \int_{-\infty}^{\infty} x^2 f(x) \, dx$$

+ Para nuestra fdp \( f(x) \):

$$E[X^2] = \int_{0}^{1} x^2 \cdot \frac{3}{2}(1 - x^2) \, dx$$

+ Realizamos esta integral:

$$E[X^2] = \frac{3}{2} \int_{0}^{1} x^2 (1 - x^2) \, dx$$

$$E[X^2] = \frac{3}{2} \left( \int_{0}^{1} x^2 \, dx - \int_{0}^{1} x^4 \, dx \right)$$

$$E[X^2] = \frac{3}{2} \left( \left[ \frac{x^3}{3} \right]_{0}^{1} - \left[ \frac{x^5}{5} \right]_{0}^{1} \right)$$

$$E[X^2] = \frac{3}{2} \left( \frac{1}{3} - \frac{1}{5} \right)$$

$$E[X^2] = \frac{3}{2} \cdot \frac{2}{15}$$

$$E[X^2] = \frac{1}{5}$$

**b. Calcular $\text{Var}(X)$:**

$$\text{Var}(X) = E[X^2] - (E[X])^2$$

$$\text{Var}(X) = \frac{1}{5} - \left( \frac{3}{8} \right)^2$$

$$\text{Var}(X) = \frac{1}{5} - \frac{9}{64}$$

$$\text{Var}(X) = \frac{64}{320} - \frac{45}{320}$$

$$\text{Var}(X) = \frac{19}{320}$$

+ **c. Calcular la desviación estándar $\sigma_X$:**

$$\sigma_X = \sqrt{\text{Var}(X)}$$

$$\sigma_X = \sqrt{\frac{19}{320}}$$

$$\sigma_X = \sqrt{\frac{19}{320}} \cdot \sqrt{\frac{320}{320}}$$

$$\sigma_X = \sqrt{\frac{19}{320}} \cdot \frac{\sqrt{320}}{320}$$

$$\sigma_X = \frac{\sqrt{19 \cdot 320}}{320}$$

$$\sigma_X = \frac{\sqrt{6080}}{320}$$
$$\sigma_X = \frac{78}{320}$$

$$\sigma_X = \frac{39}{160}$$


# **Solución del Ejercicio 2. - Usando Python**

```python
!pip install scipy
import scipy.integrate as integrate
import numpy as np

# Definir la función de densidad de probabilidad f(x)
def f(x):
    if 0 <= x <= 1:
        return 1.5 * (1 - x**2)
    else:
        return 0

# Calcular la esperanza (E[X])
def expected_value():
    expected_value, _ = integrate.quad(lambda x: x * f(x), 0, 1)
    return expected_value

# Calcular E[X^2]
def expected_value_squared():
    expected_value_squared, _ = integrate.quad(lambda x: x**2 * f(x), 0, 1)
    return expected_value_squared

# Calcular la varianza
def variance():
    ex_squared = expected_value_squared()
    ex = expected_value()
    variance = ex_squared - ex**2
    return variance

# Calcular la desviación estándar (sigma_X)
def standard_deviation():
    variance_value = variance()
    standard_deviation = np.sqrt(variance_value)
    return standard_deviation

# Calcular la esperanza, varianza y desviación estándar
mean = expected_value()
variance_value = variance()
std_deviation = standard_deviation()

# Imprimir resultados
print(f"Esperanza (E[X]): {mean:.4f} toneladas")
print(f"Varianza (Var(X)): {variance_value:.4f} toneladas^2")
print(f"Desviación estándar (sigma_X): {std_deviation:.4f} toneladas")
```

# **Ejercicio 4.**
+  **Un reloj de manecillas se detuvo en un punto que no sabemos. Determine la probabilidad de que se haya detenido en los primeros 25 minutos luego de señalar la hora en punto.**

# **Solución del Ejercicio 4.**
+ Para resolver este problema, primero debemos entender el espacio muestral y la distribución del tiempo en el reloj de manecillas.

+ **El reloj tiene una manecilla que marca los minutos y otra que marca las horas. Supongamos que el reloj se detuvo al azar en un punto desconocido entre las 12:00 y las 12:59. Esto significa que la manecilla de los minutos puede estar en cualquier posición entre 0 y 59 minutos.**

+ **El problema nos pide determinar la probabilidad de que la manecilla de los minutos se haya detenido en los primeros 25 minutos después de señalar la hora en punto (es decir, entre 0 y 25 minutos).**

+ **Pasos para resolver:**

+ 1. **Espacio Muestral**: La manecilla de los minutos puede parar en cualquiera de los 60 minutos del reloj, por lo tanto, el espacio muestral es $\Omega = \{0, 1, 2, \ldots, 59\}$

+ 2. **Distribución Uniforme**: Dado que no se menciona ninguna preferencia o sesgo hacia algún punto del intervalo de 60 minutos, asumiremos que la distribución es uniforme. Esto significa que cada minuto tiene la misma probabilidad de ser el punto de parada del reloj.

+ 3. **Probabilidad deseada**: Queremos calcular $P(0 \leq X \leq 25)$, donde $X$ es la posición de la manecilla de los minutos cuando se detiene.

+ 4. **Cálculo de la probabilidad**: La probabilidad de que la manecilla de los minutos se detenga en los primeros 25 minutos se calcula dividiendo el número de minutos deseados (0 a 25) entre el total de minutos posibles (0 a 59).

$$P(0 \leq X \leq 25) = \frac{\text{Número de minutos deseados}}{\text{Total de minutos posibles}} = \frac{26}{60}$$

+ Esto se debe a que hay 26 minutos posibles (del 0 al 25) dentro de un total de 60 minutos.

$$P(0 \leq X \leq 25) = \frac{26}{60}$$

+ Simplificando esta fracción:

$$P(0 \leq X \leq 25) = \frac{13}{30}$$

+ Por lo tanto, la probabilidad de que la manecilla de los minutos se haya detenido en los primeros 25 minutos después de señalar la hora en punto es $$\frac{13}{30}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el espacio muestral y la distribución uniforme
espacio_muestral = np.arange(0, 60)  # Posibles minutos donde se detuvo la manecilla
probabilidad_uniforme = 1 / len(espacio_muestral)  # Probabilidad uniforme para cada posición

# Calcular la probabilidad de que se detenga en los primeros 25 minutos
minutos_deseados = np.arange(0, 26)
probabilidad_deseada = len(minutos_deseados) * probabilidad_uniforme

print(f"Probabilidad de detenerse en los primeros 25 minutos: {probabilidad_deseada:.4f}")

# Crear el gráfico de la distribución uniforme
plt.figure(figsize=(10, 6))

# Barras para la probabilidad uniforme
plt.bar(espacio_muestral, np.full_like(espacio_muestral, probabilidad_uniforme), align='center', alpha=0.2, color='b', label='Probabilidad uniforme')

# Barras para la probabilidad deseada (primeros 25 minutos)
plt.bar(minutos_deseados, np.full_like(minutos_deseados, probabilidad_uniforme), align='center', color='red', label='Probabilidad deseada')

# Configuración adicional del gráfico
plt.axvspan(0, 25, alpha=0.1, color='green', label='Área de interés (0-25 minutos)')
plt.axvline(x=25.5, color='k', linestyle='--', linewidth=1, label='Límite de 25 minutos')
plt.xlabel('Minutos')
plt.ylabel('Probabilidad')
plt.title('Distribución de la posición donde se detiene la manecilla de los minutos')
plt.xticks(np.arange(0, 61, step=5))
plt.legend()
plt.grid(True)
plt.ylim(0, 1.0)
plt.xlim(0, 61)
plt.show()

```

# **Ejercicio 5.**
+ Una llamada telefónica llego a un conmutador en un tiempo, al azar, dentro de un periodo de un minuto. el conmutador estuvo ocupado durante 15 segundos en ese minuto. calcule la probabilidad de que la llamada haya llegado mientras el conmutador
no estuvo ocupado.

# **Solución del Ejercicio 4.**
+ Para resolver este problema, vamos a considerar el siguiente escenario:

+ 1. El tiempo de llegada de la llamada al conmutador es aleatorio y uniformemente distribuido dentro de un periodo de un minuto (0 a 60 segundos).
+ 2. El conmutador estuvo ocupado durante los primeros 15 segundos de ese minuto.

+ Para calcular la probabilidad de que la llamada haya llegado mientras el conmutador no estaba ocupado, debemos determinar en qué intervalo de tiempo específico debe estar la llegada de la llamada para que no coincida con el tiempo en que el conmutador estaba ocupado (0 a 15 segundos).

+ **Pasos para resolver:**

+ 1. **Espacio Muestral**: El espacio muestral para el tiempo de llegada de la llamada es $ \Omega = [0, 60]$.

+ 2. **Evento de interés**: Queremos calcular la probabilidad de que el tiempo de llegada de la llamada esté fuera del intervalo de 0 a 15 segundos.

+ 3. **Probabilidad deseada**: Esta es la probabilidad de que la llamada haya llegado en el intervalo de tiempo de 15 a 60 segundos.

+ **Cálculo de la probabilidad**:

  - El tiempo total es de 60 segundos.
  - El tiempo ocupado del conmutador es de 15 segundos.

+ Por lo tanto, el tiempo disponible (no ocupado) es de $( 60 - 15 = 45 )$ segundos.

+ La probabilidad de que la llamada haya llegado mientras el conmutador no estaba ocupado es entonces:

$$P(\text{Llamada no llegó mientras conmutador ocupado}) = \frac{\text{Tiempo no ocupado}}{\text{Tiempo total}} = \frac{45}{60} = 0.75$$

+ Es decir, hay una probabilidad del 75% de que la llamada haya llegado mientras el conmutador no estaba ocupado.

### **En Python:**

+ Vamos a visualizar esto mediante un gráfico que represente el periodo de tiempo ocupado y el periodo de tiempo no ocupado dentro del minuto:

  

```python

import matplotlib.pyplot as plt

# Definir intervalos de tiempo
tiempo_total = 60  # segundos
tiempo_ocupado = 15  # segundos

# Calcular tiempo no ocupado
tiempo_no_ocupado = tiempo_total - tiempo_ocupado

# Calcular probabilidades
probabilidad_no_ocupado = tiempo_no_ocupado / tiempo_total
probabilidad_ocupado = tiempo_ocupado / tiempo_total

# Graficar el periodo de tiempo ocupado y no ocupado
plt.figure(figsize=(8, 6))
plt.barh(0, tiempo_ocupado, height=0.3, align='center', color='red', label='Tiempo ocupado')
plt.barh(0, tiempo_no_ocupado, left=tiempo_ocupado, height=0.3, align='center', color='green', label='Tiempo no ocupado')
plt.xlim(0, tiempo_total)
plt.yticks([])
plt.xlabel('Tiempo (segundos)')
plt.title('Periodo de tiempo ocupado y no ocupado en un minuto')
plt.legend()
plt.grid(True)
plt.show()

print(f"Probabilidad de que la llamada llegue mientras el conmutador no está ocupado: {probabilidad_no_ocupado:.2f}")
```

# **Ejercicio 8.**
+  Dos amigos Roberto y Fernando, deben encontrarse en una parada de bus entre las 9:00 y las 10:00h. Cada uno esperará un máximo de 10 minutos. ¿Cuál es la probabilidad de que no se encuentren, si Fernando llegará a las 9:30 en punto?

+ **Para resolver este problema, vamos a considerar los siguientes puntos:**

+ 1. Roberto y Fernando deben encontrarse en algún momento entre las 9:00 y las 10:00h.
+ 2. Cada uno esperará un máximo de 10 minutos.
+ 3. Fernando llegará exactamente a las 9:30h.

+ Para determinar la probabilidad de que no se encuentren, consideraremos los siguientes escenarios:

 - Roberto puede llegar en cualquier momento entre las 9:00 y las 10:00h.
 - Fernando llegará exactamente a las 9:30h y esperará hasta las 9:40h (10 minutos después de su llegada).

+ **Pasos para resolver:**

+ 1. **Espacio Muestral**: El espacio muestral es el intervalo de tiempo de una hora, es decir, $\Omega = [0, 60]$ minutos.

+ 2. **Evento de interés**: Queremos calcular la probabilidad de que Roberto no llegue al mismo tiempo que Fernando (entre las 9:30 y las 9:40h).

+ 3. **Cálculo de la probabilidad**:

   - Fernando llega exactamente a las 9:30h y espera hasta las 9:40h.
   - Roberto puede llegar en cualquier momento entre las 9:00 y las 10:00h.

   - La ventana de tiempo en la que Roberto no se encontrará con Fernando es desde las 9:30h hasta las 9:40h.

   - Entonces, la probabilidad de que Roberto no llegue durante este período y, por lo tanto, no se encuentren es:

   $$
   P(\text{No se encuentran}) = \frac{\text{Tiempo que Roberto no llega}}{\text{Tiempo total}} = \frac{10}{60} = \frac{1}{6}
   $$

   + Esto es porque Roberto tiene una ventana de 10 minutos dentro de una hora total.


```python

import matplotlib.pyplot as plt

# Definir intervalo de tiempo
tiempo_total = 60  # minutos

# Tiempo máximo de espera después de la llegada de Fernando
tiempo_espera = 10  # minutos

# Tiempo en el que Fernando llega
tiempo_llegada_fernando = 30  # minutos (9:30h)

# Tiempo en el que Roberto no debe llegar para no encontrarse con Fernando
tiempo_no_encuentro_inicio = tiempo_llegada_fernando
tiempo_no_encuentro_fin = tiempo_llegada_fernando + tiempo_espera

# Calcular la probabilidad de que no se encuentren
probabilidad_no_encuentro = tiempo_espera / tiempo_total

# Crear el gráfico
plt.figure(figsize=(10, 6))

# Dibujar la línea de tiempo total
plt.plot([0, tiempo_total], [0, 0], 'k-', lw=2)

# Dibujar el período de espera de Fernando
plt.fill_between([tiempo_llegada_fernando, tiempo_no_encuentro_fin], [-1, -1], [1, 1], color='blue', alpha=0.2, label='Período de espera de Fernando')

# Configurar el gráfico
plt.xlim(0, tiempo_total)
plt.ylim(-1, 1)
plt.xlabel('Tiempo (minutos)')
plt.title('Intervalo de tiempo para encuentro entre Roberto y Fernando')
plt.legend()
plt.grid(True)
plt.show()

print(f"Probabilidad de que no se encuentren: {probabilidad_no_encuentro:.2f}")
```

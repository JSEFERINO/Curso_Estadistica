# **3. Ejercicio -Distribución Normal**
## **Problema.  El tiempo de vida de las baterías del conejito tiene una distribución aproximada a la normal con una media de 85 horas y una desviación estándar de 3.8 horas.**

## **a. Construya el gráfico de la Distribución dada si los parámetros son $μ=85,σ=3.8$**

```python
#**a. Construya el gráfico de la Distribución dada si los parámetros son μ=85,σ=3.8**
plt.figure(figsize=(8, 8))
media = 85 # parametro
sigma = 3.8 # parametro
normal = stats.norm(media,sigma )
x = np.linspace(normal.ppf(0.01),
                normal.ppf(0.99), 100)
fp = normal.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Normal')
plt.ylabel('Probabilidad')
plt.xlabel('Duración')
plt.show()
```

##**b. Construya el histograma para una muestra aleatoria tomada de una población normal  si los parámetros son $μ=85,σ=3.8$**

```python
#**b. Construya el histograma para una muestra aleatoria tomada de una población normal  si los parámetros son $μ=85,σ=3.8$**
aleatorios = stats.norm(85, 3.8).rvs(10000) # genera aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios, 10)

plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma Distribución Normal con $μ=85,σ=3.8$')
plt.show()
```

# **c. Construya la FDA para una muestra aleatoria tomada de una población normal  si los parámetros son $μ=85,σ=3.8$**

```python
# **c. Calcula la FDA para esta variable aleatoria**
x=list(range(70,100)) #El número de reacciones negativas antes de que apareciera la 5ta reacción positiva
media = 85 # parametro
sigma = 3.8 # parametro
rv=stats.norm(media, sigma)
df=pd.DataFrame({'Duración (horas)':x,'Distribución F(i)=P{X<=i}':rv.cdf(x),})
df
```

```python
#  **c. Gráfica la FDA para esta variable aleatoria**
plt.figure(figsize=(8, 8))
x = np.linspace(70, 100, 200)
distributions = pd.DataFrame({
    'x': x,
    'norm_cdf': stats.norm(loc=85, scale=3.8).cdf(x),
})
ax = distributions.plot(x='x', y='norm_cdf', legend=True, color='r')
ax.set_xlabel('$x$')
ax.set_ylabel('Standard Normal c.d.f.')
plt.title('Histograma Distribución Normal Acumulada con $μ=85,σ=3.8$')
plt.show()
```

##**e. Suponga que se tienen 3 tipos de prototipos de baterias con el mismo tiempo promedio de duración pero se ha logrado minmizar su desviación estandar a los valores de  1, 2 y 3 respectivamente. Construya en un solo cuadro las distribuciones que representan cada caso**

```python
# **e. Comparaciones de Distribuciones Normales con desviaciones diferentes**
plt.figure(figsize=(8, 8))
x = np.linspace(75, 95, 100)
distributions = pd.DataFrame({
    'x': x,
    'norm_1': stats.norm(loc=85, scale=1).pdf(x),
    'norm_2': stats.norm(loc=85, scale=2).pdf(x),
    'norm_3': stats.norm(loc=85, scale=3).pdf(x),
})
ax = distributions.plot(x='x', y='norm_1', legend=True, color='b')
distributions.plot(x='x', y='norm_2', legend=True, color='r', linestyle='--', ax=ax)
distributions.plot(x='x', y='norm_3', legend=True, color='g', linestyle='-.', ax=ax)
ax.text(82, 0.3, '$\sigma=1$')
ax.text(82, 0.2, '$\sigma=2$')
ax.text(80, 0.1, '$\sigma=3$')
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
plt.show()
```

##**f. Suponga que se tienen 3 tipos de prototipos de baterias con diferentes tiempos promedios de duración 70, 80, 90 pero se ha logrado estabilizar  su desviación estandar al valor de 3. Construya en un solo cuadro las distribuciones que representan cada caso**

```python
# **f. Comparaciones de Distribuciones Normales con medias diferentes**
plt.figure(figsize=(10, 10))
x = np.linspace(60, 100, 100)
distributions = pd.DataFrame({
    'x': x,
    'norm_1': stats.norm(loc=70, scale=3).pdf(x),
    'norm_2': stats.norm(loc=80, scale=3).pdf(x),
    'norm_3': stats.norm(loc=90, scale=3).pdf(x),
})
ax = distributions.plot(x='x', y='norm_1', legend=True, color='b')
distributions.plot(x='x', y='norm_2', legend=True, color='r', linestyle='--', ax=ax)
distributions.plot(x='x', y='norm_3', legend=True, color='g', linestyle='-.', ax=ax)
ax.text(62, 0.1, '$\mu=70$')
ax.text(75, 0.1, '$\mu=80$')
ax.text(96, 0.1, '$\mu=90$')
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
plt.show()
```

#**g. Suponga que se tienen 3 tipos de prototipos de baterias con diferentes tiempos promedios de duración 70, 80, 90 pero se ha logrado minmizar su desviación estandar a los valores de  1, 2 y 3 respectivamente. Construya en un solo cuadro las distribuciones que representan cada caso**

```python
# **g. Comparaciones de Distribuciones Normales con medias y desviaciones diferentes**
plt.figure(figsize=(10, 10))
x = np.linspace(60, 100, 100)
distributions = pd.DataFrame({
    'x': x,
    'norm_1': stats.norm(loc=70, scale=1).pdf(x),
    'norm_2': stats.norm(loc=80, scale=2).pdf(x),
    'norm_3': stats.norm(loc=90, scale=3).pdf(x),
})
ax = distributions.plot(x='x', y='norm_1', legend=True, color='b')
distributions.plot(x='x', y='norm_2', legend=True, color='r', linestyle='--', ax=ax)
distributions.plot(x='x', y='norm_3', legend=True, color='g', linestyle='-.', ax=ax)
ax.text(62, 0.1, '$\mu=70, \sigma=1$')
ax.text(75, 0.1, '$\mu=80, \sigma=2$')
ax.text(96, 0.1, '$\mu=90, \sigma=3$')
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
plt.show()
```

# **h. ¿Qué porcentaje de las baterías se espera que duren 80 horas o menos?**

```python
# **h. ¿Qué porcentaje de las baterías se espera que duren 80 horas o menos?**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(70, 100, 401)
mu, sd = 85, 3.8
y = stats.norm(mu, sd).pdf(x)
a = 80 # Valor de la variable a calcular a la izquierda

ax.plot(x, y, c='g')

#ax.fill_between(x[x>87.9], y[x>87.9], color="r")
ax.fill_between(x[x<80], y[x<80], color="b")

ax.text(75, 0.02, f'{stats.norm(mu, sd).cdf(a):.4f}', horizontalalignment='center')
#ax.text(95, 0.05, f'{1 - stats.norm(mu, sd).cdf(a):.4f}', horizontalalignment='center')

ax.set_xlabel('Valores de x')
ax.set_ylabel('Densidad')
plt.title('Histograma Distribución Normal con $μ=85,σ=3.8$')
print('El porcentaje de las baterías que se espera que duren 80 horas o menos es')
print('P{X<=80} =',round(stats.norm(loc=85, scale=3.8).cdf(80),3))

plt.show()
```

# **i. ¿Cuál es la probabilidad de que una batería dure entre 86.0 y 87.0 horas**

```python
# **i. ¿Cuál es la probabilidad de que una batería dure entre 86.0 y 87.0 horas**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(70, 100, 401)
mu, sd = 85, 3.8
y = stats.norm(mu, sd).pdf(x)
a = 80 # Valor de la variable a calcular a la izquierda
b = 87 # Valor de la variable a calcular a la izquierda
c = stats.norm(mu, sd).cdf(b)-stats.norm(mu, sd).cdf(a) # Probabilidad pedida
x1 = np.linspace(80, 87, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.norm(mu, sd).pdf(x1) , color="r")
#ax.fill_between(x[x<87], y[x<87], color="b")

ax.text(84, 0.04, f'{c:.4f}', horizontalalignment='center')
#ax.text(95, 0.05, f'{1 - stats.norm(mu, sd).cdf(a):.4f}', horizontalalignment='center')

ax.set_xlabel('Valores de x')
ax.set_ylabel('Densidad')
plt.title('Histograma Distribución Normal con $μ=85,σ=3.8$')
print('la probabilidad de que una batería dure entre 86.0 y 87.0 es')
print('P{86=<X<=80} =',c)

plt.show()
```

# **j. ¿Cuál es la probabilidad de que una batería dure más de 88 horas?**

```python
# **i. ¿Cuál es la probabilidad de que una batería dure más de 88 horas**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(70, 100, 401)
mu, sd = 85, 3.8
y = stats.norm(mu, sd).pdf(x)
#a = 80 # Valor de la variable a calcular a la izquierda
b = 88 # Valor de la variable a calcular a la izquierda
c = 1 - stats.norm(mu, sd).cdf(b) # Probabilidad pedida
x1 = np.linspace(b, 100, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.norm(mu, sd).pdf(x1) , color="r")

ax.text(90, 0.02, f'{c:.4f}', horizontalalignment='center')
#ax.text(95, 0.05, f'{1 - stats.norm(mu, sd).cdf(a):.4f}', horizontalalignment='center')

ax.set_xlabel('Valores de x')
ax.set_ylabel('Densidad')
plt.title('Histograma Distribución Normal con $μ=85,σ=3.8$')
print('la probabilidad de que una batería dure más de 88 horas es')
print('P{88=<X8} =',c)

plt.show()
```

#**k. ¿Cuál es la probabilidad de que una batería MÁS  dos desviaciones estandard de su media?**

```python
#**k. ¿Cuál es la probabilidad de que una batería dure más a lo más dos desviaciones estandard de su media?**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(70, 100, 401)
mu, sd = 85, 3.8
y = stats.norm(mu, sd).pdf(x)
a = 77.4 # Valor de la variable a calcular a la izquierda
b = 92.6 # Valor de la variable a calcular a la izquierda
c = stats.norm(mu, sd).cdf(a) # Probabilidad acumulada
d = stats.norm(mu, sd).cdf(b) # Probabilidad acumulada
x1 = np.linspace(70, a, 401)
x2 = np.linspace(b, 100, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.norm(mu, sd).pdf(x1) , color="r")
ax.fill_between(x2, 0,stats.norm(mu, sd).pdf(x2) , color="b")

ax.text(b+0.1, 0.02, f'{1-d:.4f}', horizontalalignment='center')
ax.text(a-0.1, 0.02, f'{c:.4f}', horizontalalignment='center')

ax.set_xlabel('Valores de x')
ax.set_ylabel('Densidad')
plt.title('Histograma Distribución Normal con $μ=85,σ=3.8$')
print('la probabilidad de que una batería dure más a lo más dos desviaciones estandard de su media es')
print('P{88=<X8} =',c+1-d)

plt.show()
```

# **l. Dada una distribución normal estándar con $μ=85,σ=3.8$, calcule el valor de k tal que P (X > k) = 0.3015**


```python
# **l. Dada una distribución normal estándar con $μ=85,σ=3.8$, calcule el valor de k tal que P (X > k) = 0.3015**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(70, 100, 401)
mu, sd = 85, 3.8
y = stats.norm(mu, sd).pdf(x)
b = round(stats.norm(mu, sd).ppf(0.6985),3) # valor critico pedido
#a = 92.6 # Valor de la variable a calcular a la izquierda
#c = stats.norm(mu, sd).cdf(a) # Probabilidad acumulada
d = stats.norm(mu, sd).cdf(b) # Probabilidad acumulada
#x1 = np.linspace(70, a, 401)
x2 = np.linspace(b, 100, 401)

ax.plot(x, y, c='g')

#ax.fill_between(x1, 0,stats.norm(mu, sd).pdf(x1) , color="r")
ax.fill_between(x2, 0,stats.norm(mu, sd).pdf(x2) , color="b", alpha =0.3)

ax.text(b+2, 0.02, f'{1-d:.4f}', horizontalalignment='center')
#ax.text(a-0.1, 0.02, f'{c:.4f}', horizontalalignment='center')

ax.set_xlabel('Valores de x')
ax.set_ylabel('Densidad')
plt.title('Histograma Distribución Normal con $μ=85,σ=3.8$')
print('la probabilidad de que una batería dure más a lo más dos desviaciones estandard de su media es')
print('El valor solicitado es k =',b)

plt.show()
```

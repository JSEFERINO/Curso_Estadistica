```python
# @title **Ejemplo 1: Construye la grafica de la distribución normal con media 10 y desviación 2**

import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal
media = 10
desviacion_estandar = 2

# Rango de valores para la gráfica
x = np.linspace(media - 4 * desviacion_estandar, media + 4 * desviacion_estandar, 100)

# Función de densidad de probabilidad (PDF) de la distribución normal
pdf = (1 / (desviacion_estandar * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - media) / desviacion_estandar)**2)

# Crear la gráfica
plt.plot(x, pdf)
plt.title('Distribución Normal')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
# @title **Ejemplo 2: Construye un grafico donde me dibujes curvas normales con media 10 y las diferentes desviaciones**

import numpy as np
import matplotlib.pyplot as plt

# Parámetros fijos
media = 10

# Diferentes desviaciones estándar
desviaciones_estandar = [1, 2, 3, 4]

# Rango de valores para la gráfica (ajustado para abarcar todas las desviaciones)
x = np.linspace(media - 12, media + 12, 500)

# Crear la gráfica
plt.figure(figsize=(10, 6))  # Ajustar tamaño de la figura

for desviacion_estandar in desviaciones_estandar:
    # Función de densidad de probabilidad (PDF)
    pdf = (1 / (desviacion_estandar * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - media) / desviacion_estandar)**2)
    plt.plot(x, pdf, label=f'Desviación Estándar = {desviacion_estandar}')

plt.title('Distribuciones Normales con Diferentes Desviaciones Estándar')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# @title **Ejemplo 7: Utiliza lo anterior y me hallas el área a la izquierda de 1.85. Dibuja la región y me escribes el grafico en el porcentaje**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal estándar
media = 0
desviacion_estandar = 1

# Valor para el cual se calcula el área a la izquierda
x = 1.85

# Calcular la función de distribución acumulada (CDF) para x
area_izquierda = norm.cdf(x, loc=media, scale=desviacion_estandar)

# Rango de valores para la gráfica
x_vals = np.linspace(-4, 4, 500)

# Función de densidad de probabilidad (PDF)
pdf = norm.pdf(x_vals, loc=media, scale=desviacion_estandar)

# Crear la gráfica
plt.plot(x_vals, pdf)
plt.fill_between(x_vals, pdf, where=(x_vals <= x), alpha=0.3, color='skyblue')

# Marcar el punto x en la gráfica
plt.plot(x, norm.pdf(x), marker='o', markersize=5, color='red')

# Agregar texto para mostrar el área a la izquierda
plt.text(x - 0.5, 0.1, f'P(X ≤ {x:.2f}) = {area_izquierda:.4f}', fontsize=10)

# Configurar la gráfica
plt.title('Distribución Normal Estándar')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Mostrar la gráfica
plt.show()


# Mostrar el resultado como porcentaje
porcentaje = area_izquierda * 100
print(f"El área a la izquierda de {x} es: {porcentaje:.2f}%")
```

```python
# @title **ejercicio 1 Resuelve: El volumen de hojas que una oficina de abogados deposita en un juzgados de sus casos atendidos es una variable aleatoria normal para la cual μ = 1200  y σ = 250. Determine la probabilidad de que el volumen depositado sea menor a 1050 hojas y mayor a 980. Haz el grafico, sombrea el área y coloca el porcentaje del grafico**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
mu = 1200
sigma = 250

# Valores de interés
x1 = 1050
x2 = 980

# Calcular la probabilidad de que el volumen sea menor a 1050
prob_menor_1050 = norm.cdf(x1, loc=mu, scale=sigma)

# Calcular la probabilidad de que el volumen sea mayor a 980
prob_mayor_980 = 1 - norm.cdf(x2, loc=mu, scale=sigma)

# Calcular la probabilidad de que el volumen esté entre 980 y 1050
prob_entre_980_1050 = prob_mayor_980 - (1-prob_menor_1050)


# Rango de valores para la gráfica
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 500)

# Función de densidad de probabilidad (PDF)
pdf = norm.pdf(x, loc=mu, scale=sigma)

# Crear la gráfica
plt.plot(x, pdf)

# Sombrear el área entre 980 y 1050
plt.fill_between(x, pdf, where=((x >= x2) & (x <= x1)), alpha=0.3, color='skyblue')

# Marcar los puntos x1 y x2
plt.plot(x1, norm.pdf(x1, loc=mu, scale=sigma), marker='o', markersize=5, color='red')
plt.plot(x2, norm.pdf(x2, loc=mu, scale=sigma), marker='o', markersize=5, color='red')

# Agregar texto para mostrar la probabilidad
plt.text((x1 + x2) / 2, 0.0005, f'P({x2} < X < {x1}) = {prob_entre_980_1050:.4f}', fontsize=10)

# Configurar la gráfica
plt.title('Distribución Normal del Volumen de Hojas')
plt.xlabel('Volumen de Hojas')
plt.ylabel('f(x)')
plt.grid(True)

# Mostrar la gráfica
plt.show()

print(f"La probabilidad de que el volumen sea menor a {x1} es: {prob_menor_1050:.4f}")
print(f"La probabilidad de que el volumen sea mayor a {x2} es: {prob_mayor_980:.4f}")
print(f"La probabilidad de que el volumen sea entre {x2} y {x1} es: {prob_entre_980_1050:.4f}")
```

```python
# @title **Resuelve: Sabemos que los paquetes de 1 libra de carne para hamburguesas tienen un contenido medio de 1.00 lb. con una desviación estándar de 0.13, Si tomamos al azar 50 paquetes de 1 libra de carne para hamburguesas calcule la probabilidad de que el peso medio de los paquetes sea mayor a 1.05 libras. Haz el grafico, sombrea el área y coloca el porcentaje del grafico**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros
mu = 1.00  # Media
sigma = 0.13  # Desviación estándar
n = 50  # Tamaño de la muestra
x_bar = 1.05  # Peso medio a considerar

# Desviación estándar de la media muestral
sigma_x_bar = sigma / np.sqrt(n)

# Calcular la probabilidad de que la media muestral sea mayor a 1.05
z = (x_bar - mu) / sigma_x_bar
probabilidad = 1 - norm.cdf(z)

# Generar la gráfica
x = np.linspace(mu - 3 * sigma_x_bar, mu + 3 * sigma_x_bar, 500)
pdf = norm.pdf(x, loc=mu, scale=sigma_x_bar)
plt.plot(x, pdf)

# Sombrear el área
x_fill = np.linspace(x_bar, mu + 3 * sigma_x_bar, 500)
plt.fill_between(x_fill, norm.pdf(x_fill, loc=mu, scale=sigma_x_bar), alpha=0.3, color='skyblue')

# Marcar el punto x_bar
plt.plot(x_bar, norm.pdf(x_bar, loc=mu, scale=sigma_x_bar), marker='o', markersize=5, color='red')

# Agregar texto con el porcentaje
plt.text(x_bar + 0.01, 0.05, f'{probabilidad*100:.2f}%', fontsize=10)

# Configurar la gráfica
plt.title('Distribución Normal del Peso Medio de los Paquetes')
plt.xlabel('Peso Medio (lb)')
plt.ylabel('f(x)')
plt.grid(True)

# Mostrar la gráfica
plt.show()

print(f"La probabilidad de que el peso medio de los paquetes sea mayor a {x_bar} libras es: {probabilidad:.4f} o {probabilidad*100:.2f}%")
```

```python
# prompt: Resuelve: La cantidad real de café instantáneo que una máquina vierte en frascos de 4 onzas, puede considerarse como una distribución normal que tiene una media de 4,082 onzas y una desviación estándar de 0.04. ¿Porcentaje frascos se llenan con más de 4,090 onzas serán desechados por control de calidad?. Haz el grafico, sombrea el área y coloca el porcentaje del grafico

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 4.082
desviacion_estandar = 0.04

# Valor límite para el control de calidad
valor_limite = 4.090

# Calcular la probabilidad de que la cantidad de café sea mayor a 4.090 onzas
probabilidad_mayor_4090 = 1 - norm.cdf(valor_limite, loc=media, scale=desviacion_estandar)

# Rango de valores para la gráfica
x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 500)

# Función de densidad de probabilidad (PDF)
pdf = norm.pdf(x, loc=media, scale=desviacion_estandar)

# Crear la gráfica
plt.plot(x, pdf)

# Sombrear el área
x_fill = np.linspace(valor_limite, media + 3 * desviacion_estandar, 500)
plt.fill_between(x_fill, norm.pdf(x_fill, loc=media, scale=desviacion_estandar), alpha=0.3, color='skyblue')

# Marcar el punto valor_limite
plt.plot(valor_limite, norm.pdf(valor_limite, loc=media, scale=desviacion_estandar), marker='o', markersize=5, color='red')

# Agregar texto con el porcentaje
plt.text(valor_limite + 0.001, 0.02, f'{probabilidad_mayor_4090*100:.2f}%', fontsize=10)

# Configurar la gráfica
plt.title('Distribución Normal de la Cantidad de Café')
plt.xlabel('Cantidad de Café (onzas)')
plt.ylabel('f(x)')
plt.grid(True)

# Mostrar la gráfica
plt.show()

print(f"El porcentaje de frascos que se llenan con más de {valor_limite} onzas y serán desechados es: {probabilidad_mayor_4090*100:.2f}%")
```

```python
# prompt: Resuelve: Se sabe que la prueba de auto estima sigue una distribución normal con una media de 60, con una desviación estándar de 5.83. Si se aplica una prueba de autoestima a 25 personas quienes obtienen una calificación promedio de 62.1 ¿Cuál es la probabilidad de que la media muestral sea no mas que 62.1?. Haz el grafico, sombrea el área y coloca el porcentaje del grafico

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución
mu = 60
sigma = 5.83
n = 25
x_bar = 62.1

# Desviación estándar de la media muestral
sigma_x_bar = sigma / np.sqrt(n)

# Calcular la probabilidad de que la media muestral sea menor o igual a 62.1
z = (x_bar - mu) / sigma_x_bar
probabilidad = norm.cdf(z)

# Generar la gráfica
x = np.linspace(mu - 3 * sigma_x_bar, mu + 3 * sigma_x_bar, 500)
pdf = norm.pdf(x, loc=mu, scale=sigma_x_bar)
plt.plot(x, pdf)

# Sombrear el área
x_fill = np.linspace(mu - 3 * sigma_x_bar, x_bar, 500)
plt.fill_between(x_fill, norm.pdf(x_fill, loc=mu, scale=sigma_x_bar), alpha=0.3, color='skyblue')

# Marcar el punto x_bar
plt.plot(x_bar, norm.pdf(x_bar, loc=mu, scale=sigma_x_bar), marker='o', markersize=5, color='red')

# Agregar texto con el porcentaje
plt.text(x_bar - 1, 0.03, f'{probabilidad*100:.2f}%', fontsize=10)

# Configurar la gráfica
plt.title('Distribución Normal de la Media Muestral de la Autoestima')
plt.xlabel('Media Muestral de la Autoestima')
plt.ylabel('f(x)')
plt.grid(True)

# Mostrar la gráfica
plt.show()

print(f"La probabilidad de que la media muestral sea menor o igual a {x_bar} es: {probabilidad:.4f} o {probabilidad*100:.2f}%")
```

```python
# prompt: Resuelve: Las edades de los Estudiantes del prof. Hurtado siguen una distribución Normal con Media de 18.7 y desviación de 1.3. Si tomamos una muestra de 25 estudiantes, calcula la Probabilidad de que un estudiante seleccionado al azar tenga una edad promedio mayor a 21 años. Haz el grafico, sombrea el área y coloca el porcentaje del grafico

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media_poblacional = 18.7
desviacion_estandar_poblacional = 1.3
n = 25  # Tamaño de la muestra
edad_promedio_limite = 21

# Error estándar de la media
error_estandar = desviacion_estandar_poblacional / np.sqrt(n)

# Calcular el z-score
z = (edad_promedio_limite - media_poblacional) / error_estandar

# Calcular la probabilidad
probabilidad = 1 - norm.cdf(z)

# Generar la gráfica
x = np.linspace(media_poblacional - 3 * error_estandar, media_poblacional + 3 * error_estandar, 500)
pdf = norm.pdf(x, loc=media_poblacional, scale=error_estandar)
plt.plot(x, pdf)

# Sombrear el área
x_fill = np.linspace(edad_promedio_limite, media_poblacional + 3 * error_estandar, 500)
plt.fill_between(x_fill, norm.pdf(x_fill, loc=media_poblacional, scale=error_estandar), alpha=0.3, color='skyblue')

# Marcar el punto edad_promedio_limite
plt.plot(edad_promedio_limite, norm.pdf(edad_promedio_limite, loc=media_poblacional, scale=error_estandar), marker='o', markersize=5, color='red')

# Agregar texto con el porcentaje
plt.text(edad_promedio_limite + 0.1, 0.1, f'{probabilidad*100:.2f}%', fontsize=10)

# Configurar la gráfica
plt.title('Distribución Normal de la Edad Promedio de la Muestra')
plt.xlabel('Edad Promedio')
plt.ylabel('f(x)')
plt.grid(True)

# Mostrar la gráfica
plt.show()

print(f"La probabilidad de que la edad promedio de la muestra sea mayor a {edad_promedio_limite} años es: {probabilidad:.4f} o {probabilidad*100:.2f}%")
```

# **Tema 1. Diagrama de puntos** (o **dot plot**)
## **Concepto**
+ Un **diagrama de puntos** (o **dot plot**) es una representación gráfica de datos numéricos donde cada punto representa una observación individual.
+ Este tipo de gráfico es útil para visualizar la distribución de los datos de manera simple y efectiva, especialmente cuando el conjunto de datos no es muy grande.


```python
# @title **1. Graficos de Puntos**
import matplotlib.pyplot as plt
import numpy as np

data = [161, 162, 163, 165, 165, 166, 166,
        161, 162, 163, 165, 165,
        160, 162, 163, 164, ]

# Crear un gráfico de puntos donde los repetidos se dibujen uno arriba del otro
plt.figure(figsize=(8, 2))

# Contar las ocurrencias de cada valor en los datos
unique_values, counts = np.unique(data, return_counts=True)

# Dibujar los puntos, apilando los repetidos verticalmente
for value, count in zip(unique_values, counts):
    plt.plot([value] * count, range(count), 'o', color='b')  # Apilar en el eje vertical
print(f'Consideremos el numero de puntos n = {len(data)}')
plt.yticks([])  # Ocultar eje y
plt.title("Diagrama de Puntos")
plt.xlabel("Estatura (cm)")
plt.show()
```

```python

# @title **2. Graficos de Puntos - Animacion**
data = [161, 162, 163, 165, 165, 166, 166,
        161, 162, 163, 165, 165,
        160, 162, 163, 164, ]

# Crear la figura
fig, ax = plt.subplots(figsize=(10, 4))

# Función para inicializar el gráfico
def init():
    ax.clear()
    ax.set_title("Diagrama de Puntos")
    ax.set_xlabel("Estatura (cm)")
    ax.set_yticks([])

# Función para actualizar el gráfico en cada cuadro
def animate(i):
    ax.clear()
    ax.set_title("Diagrama de Puntos")
    ax.set_xlabel("Estatura (cm)")
    ax.set_yticks([])

    # Dibujar los puntos, apilando los repetidos verticalmente
    current_data = data[:i + 1]  # Agregar más datos a medida que avanza la animación
    unique_values, counts = np.unique(current_data, return_counts=True)
    for value, count in zip(unique_values, counts):
        ax.plot([value] * count, range(count), 'o', color='b')  # Apilar en el eje vertical

# Crear la animación
ani = animation.FuncAnimation(fig, animate, frames=len(data), init_func=init, interval=500, repeat=False)

# Mostrar la animación en Jupyter Notebook
HTML(ani.to_html5_video())
```


### **Elementos clave del diagrama de puntos:**
1. **Puntos individuales**: Cada punto en el gráfico representa un valor de los datos. Si hay valores repetidos, los puntos se apilan verticalmente.
2. **Eje horizontal**: Representa los valores de la variable de interés. No siempre se usa un eje vertical, ya que la frecuencia se refleja en la acumulación de puntos.



### **Cómo describir un diagrama de puntos:**

1. **Forma**:
   - Describe la distribución general de los datos, observando si son **simétricos** o están **sesgados**.
   - Puedes observar si los datos forman una curva uniforme, si tienden hacia un lado (sesgo), o si presentan varias acumulaciones (modos).
   - Ejemplos:
     - **Simétrica**: Los puntos se distribuyen de forma similar a ambos lados del centro.
     - **Sesgo a la derecha**: Los datos tienden a concentrarse en valores bajos y hay una cola hacia los valores altos.
     - **Sesgo a la izquierda**: Lo opuesto, con más datos concentrados en los valores altos.

3. **Variabilidad**:
   - La variabilidad se refiere a qué tan dispersos o concentrados están los datos. Si los puntos están muy juntos, hay **poca variabilidad**. Si están muy dispersos a lo largo del eje, hay **mucha variabilidad**.
   - Esto se puede observar fácilmente viendo el rango de valores que cubren los puntos en el gráfico.

### **Ejemplo de interpretación:**
+ Imagina un diagrama de puntos de la estatura de estudiantes. Si ves que la mayoría de los puntos están alrededor de 170 cm, con algunos datos dispersos hacia valores más bajos o más altos, podrías decir:
- **Forma**: La distribución parece simétrica, ya que los puntos se agrupan de manera similar a ambos lados de los 170 cm.
- **Fuerza**: La mayor concentración está en el rango de 165 a 175 cm, lo que sugiere que la mayoría de los estudiantes tienen estaturas cercanas a esos valores.
- **Variabilidad**: Si los puntos están dispersos entre 150 cm y 190 cm, esto indica que hay una variabilidad moderada en las estaturas.


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Datos con alta variabilidad
data = [161, 152, 153, 175, 168, 146, 167, 160, 163, 168, 169, 162, 163, 162, 169, 164,164,
        151, 181, 170, 174, 158, 136, 197, 161, 163, 179, 168, 172, 165,
        170, 181, 170, 169, 158, 126, 197, 161, 163, 189,  170, 171]

# Crear la figura
fig, ax = plt.subplots(figsize=(10, 4))

# Función para inicializar el gráfico
def init():
    ax.clear()
    ax.set_title("Diagrama de Puntos - Mucha Variabilidad")
    ax.set_xlabel("Estatura (cm)")
    ax.set_yticks([])

# Función para actualizar el gráfico en cada cuadro
def animate(i):
    ax.clear()
    ax.set_title("Diagrama de Puntos - Mucha Variabilidad")
    ax.set_xlabel("Estatura (cm)")
    ax.set_yticks([])

    # Dibujar los puntos, apilando los repetidos verticalmente
    current_data = data[:i+1]  # Agregar más datos a medida que avanza la animación
    unique_values, counts = np.unique(current_data, return_counts=True)
    for value, count in zip(unique_values, counts):
        ax.plot([value] * count, range(count), 'o', color='b')  # Apilar en el eje vertical

# Crear la animación
ani = animation.FuncAnimation(fig, animate, frames=len(data), init_func=init, interval=300, repeat=False)

# Mostrar la animación en Jupyter Notebook
HTML(ani.to_html5_video())
```

## **16. Análisis descriptivos con tres variables: 'DEFINITIVA vs ESTRATO vs SEXO' - Intervalos de confianza con t - Student**

### **a. Intervalo de confianza con $α = 0.05$ - Es decir, 95% de confianza**

```python
import numpy as np
from scipy import stats
# Nivel de significancia
alpha = 0.05


# Asumiendo que 'count' representa el tamaño de la muestra para cada fila
df_datos7['df'] = df_datos7['count'] - 1  # Grados de libertad
df_datos7['t - critico'] = stats.t.ppf(1 - alpha/2, df_datos7['df']) # Grados de libertad

# Aplicamos la fórmula del IC con t de Student
df_datos7['IC(μ) - 95%'] = df_datos7.apply(lambda row: (
    round(row['mean'] - stats.t.ppf(1 - alpha/2, row['df']) * row['std'] / np.sqrt(row['count']), 2),
    round(row['mean'] + stats.t.ppf(1 - alpha/2, row['df']) * row['std'] / np.sqrt(row['count']), 2)
), axis=1)

df_datos7
```

### **b. Intervalo de confianza con $α = 0.02$ - Es decir, 98% de confianza**

```python
import numpy as np
from scipy import stats
# Nivel de significancia
alpha = 0.02


# Asumiendo que 'count' representa el tamaño de la muestra para cada fila
df_datos7['df'] = df_datos7['count'] - 1  # Grados de libertad
df_datos7['t - critico'] = stats.t.ppf(1 - alpha/2, df_datos7['df']) # Grados de libertad

# Aplicamos la fórmula del IC con t de Student
df_datos7['IC(μ) - 95%'] = df_datos7.apply(lambda row: (
    round(row['mean'] - stats.t.ppf(1 - alpha/2, row['df']) * row['std'] / np.sqrt(row['count']), 2),
    round(row['mean'] + stats.t.ppf(1 - alpha/2, row['df']) * row['std'] / np.sqrt(row['count']), 2)
), axis=1)

df_datos7
```

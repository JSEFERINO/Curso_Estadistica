```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
sns.color_palette(palette='CMRmap')
```

```python
# @title **Ejemplo 1: Construye la gracica de la distribucion normal con media 10 y desviacion 2**

average = 10
std = 2

x = np.linspace(average - 4 * std, average + 4 * std, 100)
y = stats.norm.pdf(x, average, std)

plt.figure(figsize=(10, 8))

sns.lineplot(x=x, y=y)
plt.title('Distribucion Normal')
plt.xlabel('X')
plt.ylabel('Probabilidad')
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.show()
```

```python
# @title **Ejemplo 2. Construye un grafico donde me dibujes las curvas normales con media 10 y las diferentes desviaciones. Utiliza seaborn para generar los graficos y ten en cuenta la celda anterior**

average = 10
stds = [1, 2, 3, 4]

plt.figure(figsize=(10, 8))

for std in stds:
  x = np.linspace(average - 4 * std, average + 4 * std, 100)
  y = stats.norm.pdf(x, average, std)
  sns.lineplot(x=x, y=y, label=f'Desviación Estándar: {std}')

plt.title('Distribuciones Normales con Diferentes Desviaciones Estándar')
plt.xlabel('X')
plt.ylabel('Probabilidad')
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.legend()
plt.show()
```

```python
# @title **Ejemplo 3. Construye un grafico donde me dibujes diferentes curvas normales con media 8, 10, 12, 15 y una desviacion de 4**

averages = [8, 10, 12, 15]
std = 4

plt.figure(figsize=(10, 8))

for average in averages:
  x = np.linspace(average - 4 * std, average + 4 * std, 100)
  y = stats.norm.pdf(x, average, std)
  sns.lineplot(x=x, y=y, label=f'Media: {average}')

plt.title('Distribuciones Normales con Diferentes Medias')
plt.xlabel('X')
plt.ylabel('Probabilidad')
plt.grid(axis='both', linestyle='--', alpha=0.2)
plt.legend()
plt.show()
```

```python
# @title **Ejemplo 4. Dibuja tres curvas normales con media 2 y desviacion 3 y muestrame en ellas el porcentaje correspondiente a una, dos, y tres desviaciones estandares, ademas de escribir el porcentaje en el grafico.**

average = 21
std = 3

x = np.linspace(average - 4 * std, average + 4 * std, 100)
y = stats.norm.pdf(x, average, std)

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x=x, y=y, ax=ax, label='Curva')

palette = sns.color_palette("CMRmap", 3)

for i in range(3):
  x1 = average - ((i + 1) * std)
  x2 = average + ((i + 1) * std)
  ax.fill_between(
      x,
      y,
      where=(x >= x1) & (x <= x2),
      color=palette[i % len(palette)],
      alpha=0.2
  )

plt.title(f'Distribución Normal (Media: {average}, Desviación Estándar: {std})')
plt.xlabel('X')
plt.ylabel('Probabilidad')
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.legend()
plt.show()
```

```python
# @title **Ejemplo 5. Agregame los porcentajes de cada region**

average = 21
std = 3

x = np.linspace(average - 4 * std, average + 4 * std, 100)
y = stats.norm.pdf(x, average, std)

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x=x, y=y, ax=ax, label='Curva')

palette = sns.color_palette("CMRmap", 3)

for i in range(4):
  x1 = average - ((i + 1) * std)
  x2 = average + ((i + 1) * std)

  area = stats.norm.cdf(x2, average, std) - stats.norm.cdf(x1, average, std)

  ax.fill_between(
      x,
      y,
      where=(x >= x1) & (x <= x2),
      color=palette[i % len(palette)],
      alpha=0.2,
      label=f'{i + 1} SD ({area * 100:.2f}%)'
  )

plt.title(f'Distribución Normal (Media: {average}, Desviación Estándar: {std})')
plt.xlabel('X')
plt.ylabel('Probabilidad')
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.legend()
plt.show()
```

```python
# @title **Ejemplo 7. Usa lo anterior y me hallas el area a la izquierda de 1.85. Dibuja la region y me escribes en el grafico el porcentaje**

average = 0
std = 1

value = 1.85

x = np.linspace(average - 4 * std, average + 4 * std, 100)

y = stats.norm.pdf(x, average, std)

area = stats.norm.cdf(value, average, std)

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x=x, y=y, ax=ax, label='Curva Normal Estándar')

ax.fill_between(
  x,
  y,
  where=(x <= value),
  alpha=0.5,
  label=f'Area <= {value}: {area * 100:.2f}%'
)

plt.title('Área bajo la curva normal estándar')
plt.xlabel('X')
plt.ylabel('Probabilidad')
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.legend()
plt.show()
```

```python
# prompt: Resuelve esto: El volumen de hojas que una oficina de abogados deposita en un juzgados de sus casos atendidos es una variable aleatoria normal para la cual μ = 1200  y σ = 250. Determine la probabilidad de que el volumen depositado sea menor a 1050 hojas y mayor a 980. Haz el grafico

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

# Datos
media = 1200
desviacion_estandar = 250

# Calcular la probabilidad de que el volumen sea menor a 1050
probabilidad_menor_1050 = stats.norm.cdf(1050, loc=media, scale=desviacion_estandar)

# Calcular la probabilidad de que el volumen sea mayor a 980
probabilidad_mayor_980 = 1 - stats.norm.cdf(980, loc=media, scale=desviacion_estandar)

# Calcular la probabilidad de que el volumen sea menor a 1050 y mayor a 980
probabilidad_entre_980_y_1050 = probabilidad_mayor_980 - (1 - probabilidad_menor_1050)

# Graficar la distribución normal
x = np.linspace(media - 4 * desviacion_estandar, media + 4 * desviacion_estandar, 100)
y = stats.norm.pdf(x, loc=media, scale=desviacion_estandar)

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x=x, y=y, ax=ax, label='Distribución Normal')

# sombrear el área entre 980 y 1050
ax.fill_between(x, y, where=(x >= 980) & (x <= 1050), color='skyblue', alpha=0.5)

# Agregar texto con la probabilidad
ax.text(1000, 0.001, f'Probabilidad entre 980 y 1050: {probabilidad_entre_980_y_1050:.4f}', fontsize=12)

plt.title('Distribución del Volumen de Hojas Depositado')
plt.xlabel('Volumen de Hojas')
plt.ylabel('Probabilidad')
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.legend()
plt.show()

print(f"La probabilidad de que el volumen depositado sea menor a 1050 es: {probabilidad_menor_1050:.4f}")
print(f"La probabilidad de que el volumen depositado sea mayor a 980 es: {probabilidad_mayor_980:.4f}")
print(f"La probabilidad de que el volumen depositado sea menor a 1050 y mayor a 980 es: {probabilidad_entre_980_y_1050:.4f}")
```

```python
# prompt: Resuelve esto como el problema anterior ademas de hacer el grafico: Sabemos que los paquetes de 1 libra de carne para hamburguesas tienen un contenido medio de 1.00 lb. con una desviación estándar de 0.13, Si tomamos al azar 50 paquetes de 1 libra de carne para hamburguesas calcule la probabilidad de que el peso medio de los paquetes sea mayor a 1.05 libras

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Datos
media_poblacional = 1.00
desviacion_estandar_poblacional = 0.13
n = 50  # Tamaño de la muestra
peso_medio_deseado = 1.05

# Calcular la desviación estándar de la media muestral
desviacion_estandar_media_muestral = desviacion_estandar_poblacional / np.sqrt(n)

# Calcular la probabilidad de que la media muestral sea mayor a 1.05
probabilidad = 1 - stats.norm.cdf(peso_medio_deseado, loc=media_poblacional, scale=desviacion_estandar_media_muestral)

# Graficar la distribución normal
x = np.linspace(media_poblacional - 4 * desviacion_estandar_media_muestral, media_poblacional + 4 * desviacion_estandar_media_muestral, 100)
y = stats.norm.pdf(x, loc=media_poblacional, scale=desviacion_estandar_media_muestral)

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x=x, y=y, ax=ax, label='Distribución Normal de la Media Muestral')

# sombrear el área donde la media muestral es mayor a 1.05
ax.fill_between(x, y, where=(x >= peso_medio_deseado), color='skyblue', alpha=0.5)

# Agregar texto con la probabilidad
ax.text(1.02, 0.05, f'Probabilidad > {peso_medio_deseado}: {probabilidad:.4f}', fontsize=12)

plt.title('Distribución del Peso Medio de los Paquetes de Carne')
plt.xlabel('Peso Medio')
plt.ylabel('Probabilidad')
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.legend()
plt.show()

print(f"La probabilidad de que el peso medio de los paquetes sea mayor a {peso_medio_deseado} es: {probabilidad:.4f}")
```

```python
# prompt: Resuelve esto como el problema anterior ademas de hacer el grafico: Una máquina despachadora de nieve está ajustada para servir un promedio de 30 ml por cono. Si la cantidad de nieve está normalmente distribuida con una varianza de 225 ml. ¿Cuál es la probabilidad de de que un cono contenga entre 25 ml y 35 ml?

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

# Datos
media = 30  # Promedio de ml por cono
varianza = 225  # Varianza de la cantidad de nieve
desviacion_estandar = np.sqrt(varianza)  # Desviación estándar

# Calcular la probabilidad de que un cono contenga entre 25 ml y 35 ml
probabilidad = stats.norm.cdf(35, loc=media, scale=desviacion_estandar) - stats.norm.cdf(25, loc=media, scale=desviacion_estandar)

# Graficar la distribución normal
x = np.linspace(media - 4 * desviacion_estandar, media + 4 * desviacion_estandar, 100)
y = stats.norm.pdf(x, loc=media, scale=desviacion_estandar)

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x=x, y=y, ax=ax, label='Distribución Normal')

# sombrear el área entre 25 ml y 35 ml
ax.fill_between(x, y, where=(x >= 25) & (x <= 35), color='skyblue', alpha=0.5)

# Agregar texto con la probabilidad
ax.text(27, 0.01, f'Probabilidad entre 25 ml y 35 ml: {probabilidad:.4f}', fontsize=12)

plt.title('Distribución de la Cantidad de Nieve por Cono')
plt.xlabel('Cantidad de Nieve (ml)')
plt.ylabel('Probabilidad')
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.legend()
plt.show()

print(f"La probabilidad de que un cono contenga entre 25 ml y 35 ml es: {probabilidad:.4f}")
```

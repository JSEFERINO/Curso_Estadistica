### **Paso 9. Halle medidas de posición para datos agrupados: Q1, Q3, D3, D8 e interprete los valores encontrados**

```python
# Función para calcular cuartiles y deciles
def calcular_cuantil(N_frac, total_frecuencia, frecuencia_acumulada, frecuencia_absoluta, intervalos):
    N_cuantil = N_frac * total_frecuencia
    indice_cuantil = next(i for i, fa in enumerate(frecuencia_acumulada) if fa >= N_cuantil)
    L_cuantil = intervalos[indice_cuantil][0]
    f_cuantil = frecuencia_absoluta[indice_cuantil]
    F_a_cuantil = frecuencia_acumulada[indice_cuantil - 1] if indice_cuantil > 0 else 0
    h_cuantil = intervalos[indice_cuantil][1] - intervalos[indice_cuantil][0]
    cuantil = L_cuantil + ((N_cuantil - F_a_cuantil) / f_cuantil) * h_cuantil
    return cuantil

# 1. Cálculo de Q1 (Primer Cuartil, N_frac = 1/4)
Q1 = calcular_cuantil(1/4, total_frecuencia, frecuencia_acumulada, frecuencia_absoluta, intervalos)
print(f"Q1 (Percentil 25): {Q1:.2f}")

# 2. Cálculo de Q3 (Tercer Cuartil, N_frac = 3/4)
Q3 = calcular_cuantil(3/4, total_frecuencia, frecuencia_acumulada, frecuencia_absoluta, intervalos)
print(f"Q3 (Percentil 75): {Q3:.2f}")

# 3. Cálculo de D3 (Tercer Decil, N_frac = 3/10)
D3 = calcular_cuantil(3/10, total_frecuencia, frecuencia_acumulada, frecuencia_absoluta, intervalos)
print(f"D3 (Percentil 30): {D3:.2f}")

# 4. Cálculo de D8 (Octavo Decil, N_frac = 8/10)
D8 = calcular_cuantil(8/10, total_frecuencia, frecuencia_acumulada, frecuencia_absoluta, intervalos)
print(f"D8 (Percentil 80):{D8:.2f}")
```

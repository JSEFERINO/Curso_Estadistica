```python
# @title **Ejemplo 1. Construye la gráfica de la distribución normal con media 10 y desviación 2**

import matplotlib.pyplot as plt
import numpy as np

# Datos de la distribución normal
mu = 10  # Media
sigma = 2  # Desviación estándar
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)  # Rango de valores en X
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2) # Distribución Normal

# Crea la gráfica
plt.plot(x, y)
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Distribución Normal (media = 10, desviación = 2)')
plt.grid(True)  # Agrega una cuadrícula
plt.show()
```

```python
# @title **Ejemplo 2. Construye un gráfico donde me dibujes las curvas normales con media 10 y las diferentes desviaciones de 2, 3, 5 y 8**

import matplotlib.pyplot as plt
import numpy as np

# Datos de la distribución normal
mu = 10  # Media
sigmas = [2, 3, 5, 8]  # Diferentes desviaciones estándar

x = np.linspace(mu - 4 * 8, mu + 4 * 8, 100)  # Rango de valores en X, usando la desviación máxima

# Crea la gráfica para cada desviación
for sigma in sigmas:
    y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    plt.plot(x, y, label=f'Desviación = {sigma}')

plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Distribuciones Normales (media = 10, diferentes desviaciones)')
plt.grid(True)  # Agrega una cuadrícula
plt.legend() # Muestra la leyenda
plt.show()
```

```python
# @title **Ejemplo 3. Construye un grafico donde me dibujes diferentes curvas normales con media 8, 10, 12 y 15 y una desviacion de 4**

import matplotlib.pyplot as plt
import numpy as np

# Datos de la distribución normal
mus = [8, 10, 12, 15]  # Diferentes medias
sigma = 4  # Desviación estándar

x = np.linspace(min(mus) - 4 * sigma, max(mus) + 4 * sigma, 100)  # Rango de valores en X

# Crea la gráfica para cada media
for mu in mus:
    y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    plt.plot(x, y, label=f'Media = {mu}')

plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Distribuciones Normales (desviación = 4, diferentes medias)')
plt.grid(True)  # Agrega una cuadrícula
plt.legend() # Muestra la leyenda
plt.show()
```

```python
# @title **Ejemplo 4. Dibuja tres curvas normales con media 21 y desviación 3, y muéstrame en ellas el porcentaje correspondiente a una desviación estándar, a dos desviaciones estándar y a tres desviaciones estándar. Escribe el porcentaje en el gráfico.**

import matplotlib.pyplot as plt
import numpy as np

# Datos de la distribución normal
mu = 21  # Media
sigma = 3  # Desviación estándar
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)  # Rango de valores en X
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)  # Distribución Normal

# Crea la gráfica
plt.plot(x, y)

# Calcula los porcentajes y las líneas verticales
percentages = [68.27, 95.45, 99.73]
for i in range(1, 4):
    x_values = [mu - i * sigma, mu + i * sigma]
    y_values = [(1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((val - mu) / sigma) ** 2) for val in x_values]

    plt.plot(x_values, y_values, marker='o', markersize=5, color='red')  # Marcadores rojos
    plt.axvline(x=x_values[0], linestyle='--', color='gray')
    plt.axvline(x=x_values[1], linestyle='--', color='gray')

    # Agrega texto con el porcentaje
    plt.text(x_values[1] + 0.5, y_values[1], f'{percentages[i-1]}%', color='red')

plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Distribución Normal (media = 21, desviación = 3)')
plt.grid(True)  # Agrega una cuadrícula
plt.show()
```

```python
# @title **Ejemplo 6: Usa lo anterior y me hallas el área a la izquierda de 1.85. Dibuja la región y me escribes en el gráfico el porcentaje.**

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Datos de la distribución normal estándar
mu = 0  # Media
sigma = 1  # Desviación estándar
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)  # Rango de valores en X
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)  # Distribución Normal

# Crea la gráfica
plt.plot(x, y, color='green')

# Valor para el cual calcular el área a la izquierda
valor = 1.85

# Calcula el área a la izquierda del valor dado
area_izquierda = norm.cdf(valor)

# Dibuja la región sombreada
x_fill = np.linspace(mu - 4 * sigma, valor, 100)
y_fill = norm.pdf(x_fill)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5)

# Agrega una línea vertical en el valor dado
plt.axvline(x=valor, color='red', linestyle='--')

# Agrega el texto con el porcentaje del área
plt.text(valor + 0.1, 0.05, f'{area_izquierda*100:.2f}%', color='red')

plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Área bajo la curva normal estándar')
plt.grid(True)
plt.show()
```

```python
# @title **Ejemplo 1. Construye la gráfica de la distribución normal con media 10 y desviación 2**

import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal
mu = 10  # Media
sigma = 2  # Desviación estándar

# Rango de valores para la gráfica
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# Función de densidad de probabilidad (PDF) de la distribución normal
y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Crear la gráfica
plt.plot(x, y)

# Agregar etiquetas y título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Distribución Normal (Media = 10, Desviación Estándar = 2)')

# Agregar la línea de la media
plt.axvline(mu, color='pink', linestyle='dashed', linewidth=1, label='Media')

# Mostrar la leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
# @title **Ejemplo 2. construye un gráfico donde me dibujes las curvas normales con media 10 y las diferentes desviaciones**

import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal
mu = 10  # Media
sigmas = [1, 2, 3]  # Diferentes desviaciones estándar

# Rango de valores para la gráfica
x = np.linspace(mu - 12, mu + 12, 100) #Ajustado el rango para visualizar mejor todas las curvas

# Crear la gráfica
for sigma in sigmas:
  y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
  plt.plot(x, y, label=f'Desviación Estándar = {sigma}')

# Agregar etiquetas y título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Distribuciones Normales (Media = 10)')

# Agregar la línea de la media
plt.axvline(mu, color='red', linestyle='dashed', linewidth=1, label='Media')

# Mostrar la leyenda
plt.legend()

# Mostrar la gráfica
plt.grid(True) # Añadido grid para mejor visualización
plt.show()
```

```python
# @title **Ejemplo 3. construye un grafico donde me dibujes diferentes curvas normales con media 8, 10, 12 y 15 y desviación de 4**

import numpy as np
import matplotlib.pyplot as plt

# Medias y desviación estándar
means = [8, 10, 12, 15]
sigma = 4

# Rango de valores para la gráfica (ajustado para visualizar todas las curvas)
x = np.linspace(min(means) - 4*sigma, max(means) + 4*sigma, 500)

# Crear la gráfica
for mu in means:
    y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
    plt.plot(x, y, label=f'Media = {mu}, Desviación Estándar = {sigma}')

# Agregar etiquetas y título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Distribuciones Normales')

# Agregar líneas de las medias
for mu in means:
    plt.axvline(mu, color='red', linestyle='dashed', linewidth=1, alpha=0.7)

# Mostrar la leyenda
plt.legend()

# Agregar una cuadrícula
plt.grid(True, alpha=0.5)

# Mostrar la gráfica
plt.show()
```

```python
# @title **Ejemplo 4. Dibuja tres curvas normales con media 21 y desviación 3 y muestrame en ellas el porcentaje correspondiente a una desviacion estandar, a dos desviacion estandard y a tres desviación estandard y escriba el porcentaje en el gráfico**

import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal
mu = 21  # Media
sigma = 3  # Desviación estándar

# Rango de valores para la gráfica
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)

# Crear la gráfica para las tres curvas
for i in range(3):
    y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
    plt.plot(x, y, label=f'Curva {i+1}')

    # Añadir líneas para 1, 2 y 3 desviaciones estándar
    x_1std_left = mu - sigma
    x_1std_right = mu + sigma
    x_2std_left = mu - 2 * sigma
    x_2std_right = mu + 2 * sigma
    x_3std_left = mu - 3 * sigma
    x_3std_right = mu + 3 * sigma

    plt.axvline(x_1std_left, color='gray', linestyle='--', linewidth=1)
    plt.axvline(x_1std_right, color='gray', linestyle='--', linewidth=1)
    plt.axvline(x_2std_left, color='gray', linestyle='-.', linewidth=1)
    plt.axvline(x_2std_right, color='gray', linestyle='-.', linewidth=1)
    plt.axvline(x_3std_left, color='gray', linestyle=':', linewidth=1)
    plt.axvline(x_3std_right, color='gray', linestyle=':', linewidth=1)

    # Agregar texto de los porcentajes
    plt.text(x_1std_right + 0.5, 0.05, '68.27%', fontsize=9)
    plt.text(x_2std_right + 0.5, 0.02, '95.45%', fontsize=9)
    plt.text(x_3std_right + 0.5, 0.01, '99.73%', fontsize=9)

# Agregar etiquetas y título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Tres Distribuciones Normales (Media = 21, Desviación Estándar = 3)')

# Mostrar la leyenda
plt.legend()

# Mostrar la gráfica
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# @title **Ejemplo 7. usa lo anterior y halla el área a la izquierda de 1.85. sombrea  la región y me escribes en el gráfico el porcentaje**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
mu = 0  # Media (asumiendo una distribución estándar)
sigma = 1  # Desviación estándar (asumiendo una distribución estándar)

# Rango de valores para la gráfica
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)

# Función de densidad de probabilidad (PDF) de la distribución normal
y = norm.pdf(x, mu, sigma)

# Crear la gráfica
plt.plot(x, y)

# Valor para el cual se calcula el área a la izquierda
x_value = 1.85

# Calcular el área bajo la curva a la izquierda de x_value
area = norm.cdf(x_value, mu, sigma)

# Sombrear la región bajo la curva a la izquierda de x_value
x_fill = np.linspace(mu - 4*sigma, x_value, 500)
y_fill = norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5)


# Agregar una línea vertical en x_value
plt.axvline(x_value, color='red', linestyle='--', linewidth=1)

# Agregar el porcentaje en la gráfica
plt.text(x_value + 0.1, 0.1, f'{area*100:.2f}%', fontsize=12, color='red') #Ajustado la posición y formato


# Agregar etiquetas y título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f'Distribución Normal Estándar (Área a la izquierda de {x_value})')


# Mostrar la gráfica
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# @title **Ejercicio 1. Resuelve: El volumen de hojas que una oficina de abogados deposita en un juzgados de sus casos atendidos es una variable aleatoria normal para la cual μ = 1200  y σ = 250. Determine la probabilidad de que el volumen depositado sea menor a 1050 hojas y mayor a 980. Haz el gráfico, sombrea el área y coloca le porcentaje en el gráfico**

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
mu = 1200  # Media
sigma = 250  # Desviación estándar

# Valores límite
x1 = 980
x2 = 1050

# Rango de valores para la gráfica
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)

# Función de densidad de probabilidad (PDF) de la distribución normal
y = norm.pdf(x, mu, sigma)

# Crear la gráfica
plt.plot(x, y)

# Calcular las probabilidades
p1 = norm.cdf(x1, mu, sigma) # Probabilidad de que el volumen sea menor a 980
p2 = norm.cdf(x2, mu, sigma) # Probabilidad de que el volumen sea menor a 1050
prob = p2 - p1 # Probabilidad de que el volumen sea mayor a 980 y menor a 1050


# Sombrear la región bajo la curva entre x1 y x2
x_fill = np.linspace(x1, x2, 500)
y_fill = norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5)

# Agregar líneas verticales en x1 y x2
plt.axvline(x1, color='red', linestyle='--', linewidth=1)
plt.axvline(x2, color='red', linestyle='--', linewidth=1)

# Agregar el porcentaje en la gráfica
plt.text((x1 + x2) / 2, 0.0005, f'{prob*100:.2f}%', fontsize=12, color='red', ha='center')


# Agregar etiquetas y título
plt.xlabel('Volumen de hojas')
plt.ylabel('f(x)')
plt.title(f'Distribución Normal (Media = {mu}, Desviación Estándar = {sigma})')

# Mostrar la gráfica
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# @title **Ejercicio 2. Resuelve como el problema anterior: Sabemos que los paquetes de 1 libra de carne para hamburguesas tienen un contenido medio de 1.00 lb. con una desviación estándar de 0.13, Si tomamos al azar 50 paquetes de 1 libra de carne para hamburguesas calcule la probabilidad de que el peso medio de los paquetes sea mayor a 1.05 libras**

# Parámetros de la distribución
mu = 1.00  # Media del peso de los paquetes
sigma = 0.13  # Desviación estándar del peso de los paquetes
n = 50  # Número de paquetes en la muestra
x_bar = 1.05 # Media muestral

# Calcular la desviación estándar de la distribución muestral de medias
sigma_x_bar = sigma / np.sqrt(n)

# Calcular la probabilidad de que la media muestral sea mayor a 1.05
probabilidad = 1 - norm.cdf(x_bar, loc=mu, scale=sigma_x_bar)


# Imprimir el resultado
print(f"La probabilidad de que el peso medio de los 50 paquetes sea mayor a 1.05 libras es: {probabilidad:.4f}")


# Crear la gráfica de la distribución normal
x = np.linspace(mu - 4*sigma_x_bar, mu + 4*sigma_x_bar, 500)
y = norm.pdf(x, mu, sigma_x_bar)
plt.plot(x, y)

# Sombrear el área a la derecha de x_bar
x_fill = np.linspace(x_bar, mu + 4 * sigma_x_bar, 500)
y_fill = norm.pdf(x_fill, mu, sigma_x_bar)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5)

# Agregar una línea vertical en x_bar
plt.axvline(x_bar, color='red', linestyle='--', linewidth=1)

# Agregar el porcentaje en la gráfica
plt.text(x_bar + 0.01, 0.1, f'{probabilidad*100:.2f}%', fontsize=12, color='red')

# Agregar etiquetas y título
plt.xlabel('Peso medio de los paquetes (libras)')
plt.ylabel('f(x)')
plt.title(f'Distribución Normal (Media = {mu}, Desviación Estándar = {sigma_x_bar:.4f})')

# Mostrar la gráfica
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# @title **Ejercicio 3. Resuelve como el problema anterior: Una máquina despachadora de nieve está ajustada para servir un promedio de 30 ml por cono. Si la cantidad de nieve está normalmente distribuida con una varianza de 225 ml. ¿Cuál es la probabilidad de de que un cono contenga entre 25 ml y 35 ml?**

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Parámetros de la distribución
mu = 30  # Media
sigma = np.sqrt(225)  # Desviación estándar (raíz cuadrada de la varianza)

# Valores límite
x1 = 25
x2 = 35

# Calcular la probabilidad
prob = norm.cdf(x2, loc=mu, scale=sigma) - norm.cdf(x1, loc=mu, scale=sigma)

# Imprimir el resultado
print(f"La probabilidad de que un cono contenga entre 25 ml y 35 ml es: {prob:.4f}")

# Crear la gráfica de la distribución normal
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
y = norm.pdf(x, mu, sigma)
plt.plot(x, y)

# Sombrear el área entre x1 y x2
x_fill = np.linspace(x1, x2, 500)
y_fill = norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5)

# Agregar líneas verticales en x1 y x2
plt.axvline(x1, color='red', linestyle='--', linewidth=1)
plt.axvline(x2, color='red', linestyle='--', linewidth=1)

# Agregar el porcentaje en la gráfica
plt.text((x1 + x2) / 2, 0.005, f'{prob*100:.2f}%', fontsize=12, color='red', ha='center')

# Agregar etiquetas y título
plt.xlabel('Cantidad de nieve (ml)')
plt.ylabel('f(x)')
plt.title(f'Distribución Normal (Media = {mu}, Desviación Estándar = {sigma})')

# Mostrar la gráfica
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# @title **Ejercicio 6. Resuelve como el problema anterior: Se sabe que la prueba de auto estima sigue una distribución normal con una media de 60, con una desviación estándar de 5.83. Si se aplica una prueba de autoestima a 25 personas quienes obtienen una calificación promedio de 62.1 ¿Cuál es la probabilidad de que la media muestral sea no mas que 62.1?**

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Parámetros de la distribución
mu = 60  # Media de la población
sigma = 5.83  # Desviación estándar de la población
n = 25  # Tamaño de la muestra
x_bar = 62.1  # Media muestral

# Calcular la desviación estándar de la distribución muestral de medias
sigma_x_bar = sigma / np.sqrt(n)

# Calcular la probabilidad de que la media muestral sea no más que 62.1
probabilidad = norm.cdf(x_bar, loc=mu, scale=sigma_x_bar)

# Imprimir el resultado
print(f"La probabilidad de que la media muestral sea no más de 62.1 es: {probabilidad:.4f}")

# Crear la gráfica de la distribución normal
x = np.linspace(mu - 4*sigma_x_bar, mu + 4*sigma_x_bar, 500)
y = norm.pdf(x, mu, sigma_x_bar)
plt.plot(x, y)

# Sombrear el área a la izquierda de x_bar
x_fill = np.linspace(mu - 4*sigma_x_bar, x_bar, 500)
y_fill = norm.pdf(x_fill, mu, sigma_x_bar)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5)

# Agregar una línea vertical en x_bar
plt.axvline(x_bar, color='red', linestyle='--', linewidth=1)

# Agregar el porcentaje en la gráfica
plt.text(x_bar - 1, 0.02, f'{probabilidad*100:.2f}%', fontsize=12, color='red')

# Agregar etiquetas y título
plt.xlabel('Media muestral de la prueba de autoestima')
plt.ylabel('f(x)')
plt.title(f'Distribución Normal (Media = {mu}, Desviación Estándar = {sigma_x_bar:.4f})')

# Mostrar la gráfica
plt.grid(True, alpha=0.3)
plt.show()
```
